```python
# ==============================================================================
# 完整整合代碼：MOO- (Threshold x 優化) vs Pro (u, v 優化)
# 功能：
#   1. MOO- 策略：SA 找出最佳閾值倍率 x
#   2. Pro 策略：SA 找出最佳參數組合 (u, v)
#   3. 比較兩者在不同偏好權重下的表現
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time', 'task_count']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0
        self.task_count = 0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time
        self.task_count += 1

# ==========================================
# 2. 模擬器 (Simulator)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='moo_minus', pre_generated_tasks=None,
                 var_threshold=float('inf'), u=3, v=3):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold
        # Pro 策略參數
        self.u = int(u)
        self.v = int(v)

    def _dispatch_moo_minus(self, task, threshold_v):
        """ MOO- 策略: Constraint + Minimize(Sum_Var) """
        task_load = task.actual_time
        current_loads = np.array([m.total_load for m in self.machines])
        candidates = []

        # 階段 1: 硬約束過濾
        for i, machine in enumerate(self.machines):
            hypothetical_loads = current_loads.copy()
            hypothetical_loads[i] += task_load
            hyp_var = np.var(hypothetical_loads)
            if hyp_var <= threshold_v:
                candidates.append(machine)

        # 階段 2: 選擇
        if not candidates:
            chosen_machine = min(self.machines, key=lambda m: m.sum_mean_time)
        else:
            chosen_machine = min(candidates, key=lambda m: m.sum_var_time)
        return chosen_machine

    def _dispatch_pro(self):
        """ Pro 策略: Union(Top-u Mean, Top-v Var) -> Least Task Count """
        # 1. Mean 最小的前 u 名
        U_prime = sorted(self.machines, key=lambda m: m.sum_mean_time)[:self.u]
        # 2. Variance 最小的前 v 名
        V_prime = sorted(self.machines, key=lambda m: m.sum_var_time)[:self.v]
        # 3. 取聯集
        candidates = list(set(U_prime + V_prime))
        # 4. 選擇 task_count 最小的
        chosen_machine = min(candidates, key=lambda m: m.task_count)
        return chosen_machine

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'moo_minus':
                best_m = self._dispatch_moo_minus(task, self.var_threshold)
            elif self.strategy == 'pro':
                best_m = self._dispatch_pro()

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))
        return assignment_map

# ==========================================
# 3. 優化器 A: MOO- Threshold 優化 (找最佳 x)
# ==========================================

class Threshold_X_Optimizer:
    def __init__(self, task_pool, num_machines, target_preference_weight, baseline_var, risk_magnifier=8.0):
        self.task_pool = task_pool
        self.num_machines = num_machines
        self.target_w = target_preference_weight
        self.baseline_var = baseline_var
        self.risk_magnifier = risk_magnifier

    def calculate_objective(self, x_multiplier):
        current_threshold = self.baseline_var * x_multiplier
        sim = Simulator(self.num_machines, strategy='moo_minus',
                        pre_generated_tasks=self.task_pool,
                        var_threshold=current_threshold)
        _ = sim.run()

        m_means = np.array([m.sum_mean_time for m in sim.machines])
        m_vars = np.array([m.sum_var_time for m in sim.machines])

        mean_of_means = np.mean(m_means)
        mean_of_vars = np.mean(m_vars)
        cv_load = np.std(m_means) / mean_of_means if mean_of_means > 0 else 0
        cv_risk = np.std(m_vars) / mean_of_vars if mean_of_vars > 0 else 0

        exp_risk_penalty = np.exp(cv_risk * self.risk_magnifier) - 1.0
        weighted_cost = self.target_w * cv_load + (1 - self.target_w) * exp_risk_penalty
        return weighted_cost

    def optimize(self, max_iterations=30, init_temp=1.0):
        curr_x = 1.2
        curr_cost = self.calculate_objective(curr_x)
        best_x, best_cost = curr_x, curr_cost
        temp = init_temp

        for i in range(max_iterations):
            perturbation = np.random.normal(0, 0.2 * temp)
            new_x = np.clip(curr_x + perturbation, 0.5, 8.0)
            new_cost = self.calculate_objective(new_x)

            if new_cost < curr_cost or np.random.random() < np.exp(-(new_cost - curr_cost) / temp):
                curr_x = new_x
                curr_cost = new_cost
                if curr_cost < best_cost:
                    best_x, best_cost = new_x, new_cost
            temp *= 0.9
        return best_x, best_cost

# ==========================================
# 4. 優化器 B: Pro Strategy UV 優化 (找最佳 u, v)
# ==========================================

class Pro_UV_Optimizer:
    def __init__(self, task_pool, num_machines, target_preference_weight, risk_magnifier=8.0):
        self.task_pool = task_pool
        self.num_machines = num_machines
        self.target_w = target_preference_weight
        self.risk_magnifier = risk_magnifier

    def calculate_objective(self, u, v):
        sim = Simulator(self.num_machines, strategy='pro',
                        pre_generated_tasks=self.task_pool,
                        u=u, v=v)
        _ = sim.run()

        m_means = np.array([m.sum_mean_time for m in sim.machines])
        m_vars = np.array([m.sum_var_time for m in sim.machines])

        mean_of_means = np.mean(m_means)
        mean_of_vars = np.mean(m_vars)
        cv_load = np.std(m_means) / mean_of_means if mean_of_means > 0 else 0
        cv_risk = np.std(m_vars) / mean_of_vars if mean_of_vars > 0 else 0

        # 使用相同的 Cost Function 以便公平比較
        # exp_risk_penalty = np.exp(cv_risk * self.risk_magnifier) - 1.0
        weighted_cost = self.target_w * cv_load + (1 - self.target_w) * cv_risk
        return weighted_cost

    def optimize(self, max_iterations=30, init_temp=10.0):
        # 初始狀態：從小範圍開始搜索
        curr_u, curr_v = 3, 3
        curr_cost = self.calculate_objective(curr_u, curr_v)
        best_uv, best_cost = (curr_u, curr_v), curr_cost
        temp = init_temp

        # 因為 u, v 是整數，我們使用隨機整數擾動
        for i in range(max_iterations):
            # 產生鄰居：隨機加減 1~2
            du = np.random.randint(-2, 3)
            dv = np.random.randint(-2, 3)

            new_u = int(np.clip(curr_u + du, 1, self.num_machines))
            new_v = int(np.clip(curr_v + dv, 1, self.num_machines))

            new_cost = self.calculate_objective(new_u, new_v)

            if new_cost < curr_cost or np.random.random() < np.exp(-(new_cost - curr_cost) / (temp + 1e-9)):
                curr_u, curr_v = new_u, new_v
                curr_cost = new_cost
                if curr_cost < best_cost:
                    best_uv, best_cost = (new_u, new_v), new_cost

            temp *= 0.9

        return best_uv, best_cost

# ==========================================
# 5. GPU 驗證與繪圖
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    print(f"\n🚀 開始 Monte Carlo 驗證 (Runs={num_runs})...")

    for name, config in strategies_config.items():
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Load Balance CDF
    ax1 = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys(), key=lambda x: 0 if 'Baseline' in x else 1)
    colors = plt.cm.viridis(np.linspace(0, 0.9, len(results)))

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8

        ax1.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax1.set_title('Load Balance CDF (Normalized)')
    ax1.set_xlim(0.9, 1.1)
    ax1.legend(fontsize='small'); ax1.grid(True, alpha=0.3)

    # Risk CDF
    ax2 = axes[1]
    base_var_key = next((k for k in results if 'Least Mu' in k), list(results.keys())[0])
    base_var_val = np.mean(results[base_var_key]['var_loads'])

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['var_loads']) / base_var_val
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8

        ax2.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax2.set_title('Risk CDF (System Variance)')
    ax2.legend(fontsize='small'); ax2.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_pareto_scatter(results):
    fig, ax = plt.subplots(figsize=(9, 6))
    points = []

    for name, data in results.items():
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

        if 'Baseline' in name:
            color = 'gray'; marker = 'X'; size = 150; w_val = -1; label = "Baseline"
        elif 'Pro' in name:
            color = 'red'; marker = '*'; size = 250; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        elif 'MOO' in name:
            color = 'blue'; marker = 'o'; size = 150; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        else:
            w_val = 0.5; color = 'green'; marker = '^'; size = 100; label = name

        points.append({'x': cv_load, 'y': cv_risk, 'c': color, 'm': marker, 's': size, 'label': label, 'w': w_val})

    for p in points:
        ax.scatter(p['x'], p['y'], c=[p['c']], s=p['s'], marker=p['m'], edgecolors='k', zorder=5)
        offset_y = 0.0005
        ax.text(p['x'], p['y'] + offset_y, p['label'], fontsize=9, ha='center', va='bottom')

    # 畫 MOO- 的 Pareto 線
    moo_points = sorted([p for p in points if 'MOO' in p['label']], key=lambda x: x['w'])
    if moo_points:
        px = [p['x'] for p in moo_points]
        py = [p['y'] for p in moo_points]
        ax.plot(px, py, 'b--', alpha=0.4, label='MOO- Frontier')

    # 畫 Pro 的 Pareto 線
    pro_points = sorted([p for p in points if 'Pro' in p['label']], key=lambda x: x['w'])
    if pro_points:
        px = [p['x'] for p in pro_points]
        py = [p['y'] for p in pro_points]
        ax.plot(px, py, 'r-.', alpha=0.4, label='Pro Frontier')

    ax.set_title('Strategy Comparison: MOO- vs Pro (Optimized)')
    ax.set_xlabel('Load Imbalance (CV) $\\leftarrow$ Better')
    ax.set_ylabel('Risk Imbalance (CV) $\\leftarrow$ Better')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend()
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    NUM_TASKS = 10000
    NUM_MACHINES = 100
    NUM_RUNS = 10000

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 6.0
    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print("\n2. 計算 Baseline...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline System Var: {BASELINE_VAR:.2f}")

    preference_weights = [0, 0.3, 0.5, 0.7, 1.0]

    # 儲存最佳參數
    best_moo_params = {}
    best_pro_params = {}

    print("\n3. 開始雙軌優化 (MOO- vs Pro)...")
    pbar = tqdm(preference_weights, desc="Optimizing Strategies")

    for w in pbar:
        # A. 優化 MOO- (Threshold x)
        opt_moo = Threshold_X_Optimizer(task_pool, NUM_MACHINES, w, BASELINE_VAR, risk_magnifier=8.0)
        best_x, _ = opt_moo.optimize(max_iterations=30)
        best_moo_params[w] = best_x

        # B. 優化 Pro (u, v)
        opt_pro = Pro_UV_Optimizer(task_pool, NUM_MACHINES, w, risk_magnifier=8.0)
        best_uv, _ = opt_pro.optimize(max_iterations=30)
        best_pro_params[w] = best_uv

        # tqdm 顯示當前進度
        pbar.set_postfix({'W': w, 'x': f"{best_x:.2f}", 'uv': f"{best_uv}"})

    print("\n4. 執行評估 & 繪圖...")

    configs = {
        'Baseline (Least Mu)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}},
    }

    # 加入 MOO- 結果
    for w, x_val in best_moo_params.items():
        label = f'MOO- W={w} (x={x_val:.2f})'
        configs[label] = {
            'strategy': 'moo_minus',
            'params': {'var_threshold': BASELINE_VAR * x_val}
        }

    # 加入 Pro 結果
    for w, (u_val, v_val) in best_pro_params.items():
        label = f'Pro W={w} (u={u_val},v={v_val})'
        configs[label] = {
            'strategy': 'pro',
            'params': {'u': u_val, 'v': v_val}
        }

    eval_results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)

    plot_cdf(eval_results, mean_times, NUM_MACHINES)
    plot_pareto_scatter(eval_results)

    print("\n✅ 優化完成！比較圖表已生成。")
```

```python
# ==============================================================================
# 完整整合代碼：MOO- vs Pro (SA 優化版 + Numpy 極速加速)
# 特點：
#   1. SA 搜尋過程：使用 Numpy Array 取代物件操作 (CPU 極速優化)
#   2. 最終驗證：使用 CuPy 進行大規模 Monte Carlo (GPU 加速)
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time', 'task_count']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0
        self.task_count = 0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time
        self.task_count += 1

# ==========================================
# 2. 快速模擬器 (FastSimulator for SA)
# 用於 SA 頻繁呼叫，移除物件操作，改用 Numpy Array
# ==========================================

class FastSimulator:
    def __init__(self, num_machines, strategy, tasks_data, var_threshold=float('inf'), u=3, v=3):
        self.num_machines = num_machines
        self.strategy = strategy
        # tasks_data: List of (mean, var, actual) tuples
        self.tasks_data = tasks_data
        self.var_threshold = var_threshold
        self.u = int(u)
        self.v = int(v)

        # [State Arrays]
        # 0: total_load (actual/mean), 1: sum_mean, 2: sum_var, 3: count
        self.state = np.zeros((4, num_machines), dtype=np.float64)

    def run(self):
        # 提取狀態指針以加速
        s_load = self.state[0]
        s_mean = self.state[1]
        s_var = self.state[2]
        s_count = self.state[3]

        num_m = self.num_machines
        strategy = self.strategy
        threshold = self.var_threshold
        u, v = self.u, self.v

        # 預先生成索引陣列
        indices = np.arange(num_m)

        for mean_t, var_t, actual_t in self.tasks_data:
            chosen_idx = -1

            if strategy == 'moo_minus':
                # 計算加入後的變異數 (Vectorized)
                # 這裡需要稍微小心，np.var 的計算成本
                # Var(X) = E[X^2] - (E[X])^2
                # 為了極速，我們只計算當前 Load 的變異數變化

                # 複製當前 loads 並加上 task_load (Broadcasting)
                # hypothetical_loads = s_load + (mask * actual_t) -> 慢
                # 我們用迴圈檢查或數學推導

                # 為了效能，這裡我們只檢查「合格者」，不複製整個 Array
                # 1. 算出當前 sum_sq 和 sum
                current_sum = np.sum(s_load)
                current_sq_sum = np.sum(s_load**2)

                # 2. 快速推算加入某機器後的 Var
                # New_Var = (New_Sq_Sum / N) - (New_Mean)^2
                # New_Sq_Sum = Current_Sq_Sum - old^2 + (old+task)^2
                # New_Sum = Current_Sum + task

                new_sums = current_sum + actual_t
                new_mean = new_sums / num_m

                new_sq_sums = current_sq_sum - (s_load**2) + ((s_load + actual_t)**2)
                new_vars = (new_sq_sums / num_m) - (new_mean**2)

                # 篩選
                mask = new_vars <= threshold
                candidates = indices[mask]

                if len(candidates) == 0:
                    # Fallback: Least Mean
                    chosen_idx = np.argmin(s_mean)
                else:
                    # Optimize: Minimize Sum Var
                    # 在 candidates 中找 s_var 最小的
                    cand_vars = s_var[candidates]
                    best_cand_idx = np.argmin(cand_vars)
                    chosen_idx = candidates[best_cand_idx]

            elif strategy == 'pro':
                # 1. Top-u Mean (Smallest)
                # argpartition 比 sort 快
                if u >= num_m: idx_u = indices
                else: idx_u = np.argpartition(s_mean, u)[:u]

                # 2. Top-v Var (Smallest)
                if v >= num_m: idx_v = indices
                else: idx_v = np.argpartition(s_var, v)[:v]

                # 3. Union
                # Numpy 的 union1d 會排序，這裡用 unique 比較快
                candidates = np.unique(np.concatenate((idx_u, idx_v)))

                # 4. Least Task Count
                cand_counts = s_count[candidates]
                # 若 count 一樣，argmin 會回傳第一個，這可能會導致偏差，
                # 但為了速度我們接受 (或可加入隨機擾動)
                best_cand_idx = np.argmin(cand_counts)
                chosen_idx = candidates[best_cand_idx]

            # Update State
            s_load[chosen_idx] += actual_t
            s_mean[chosen_idx] += mean_t
            s_var[chosen_idx] += var_t
            s_count[chosen_idx] += 1

        return self.state

# ==========================================
# 3. 一般模擬器 (用於最終 Monte Carlo)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='moo_minus', pre_generated_tasks=None,
                 var_threshold=float('inf'), u=3, v=3):
        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold
        self.u = int(u)
        self.v = int(v)

    def _dispatch_moo_minus(self, task, threshold_v):
        task_load = task.actual_time
        current_loads = np.array([m.total_load for m in self.machines])
        candidates = []
        for i, machine in enumerate(self.machines):
            hypothetical_loads = current_loads.copy()
            hypothetical_loads[i] += task_load
            hyp_var = np.var(hypothetical_loads)
            if hyp_var <= threshold_v:
                candidates.append(machine)
        if not candidates:
            chosen_machine = min(self.machines, key=lambda m: m.sum_mean_time)
        else:
            chosen_machine = min(candidates, key=lambda m: m.sum_var_time)
        return chosen_machine

    def _dispatch_pro(self):
        U_prime = sorted(self.machines, key=lambda m: m.sum_mean_time)[:self.u]
        V_prime = sorted(self.machines, key=lambda m: m.sum_var_time)[:self.v]
        candidates = list(set(U_prime + V_prime))
        chosen_machine = min(candidates, key=lambda m: m.task_count)
        return chosen_machine

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'moo_minus':
                best_m = self._dispatch_moo_minus(task, self.var_threshold)
            elif self.strategy == 'pro':
                best_m = self._dispatch_pro()
            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))
        return assignment_map

# ==========================================
# 4. 優化器 (使用 FastSimulator)
# ==========================================

class Threshold_X_Optimizer:
    def __init__(self, tasks_data, num_machines, target_preference_weight, baseline_var, risk_magnifier=8.0):
        self.tasks_data = tasks_data # Raw list of tuples
        self.num_machines = num_machines
        self.target_w = target_preference_weight
        self.baseline_var = baseline_var
        self.risk_magnifier = risk_magnifier

    def calculate_objective(self, x_multiplier):
        current_threshold = self.baseline_var * x_multiplier
        # 使用 FastSimulator
        sim = FastSimulator(self.num_machines, 'moo_minus', self.tasks_data, var_threshold=current_threshold)
        final_state = sim.run()

        m_means = final_state[1]
        m_vars = final_state[2]

        mean_of_means = np.mean(m_means)
        mean_of_vars = np.mean(m_vars)
        cv_load = np.std(m_means) / mean_of_means if mean_of_means > 0 else 0
        cv_risk = np.std(m_vars) / mean_of_vars if mean_of_vars > 0 else 0

        exp_risk_penalty = np.exp(cv_risk * self.risk_magnifier) - 1.0
        weighted_cost = self.target_w * cv_load + (1 - self.target_w) * exp_risk_penalty
        return weighted_cost

    def optimize(self, max_iterations=30, init_temp=1.0):
        curr_x = 1.2
        curr_cost = self.calculate_objective(curr_x)
        best_x, best_cost = curr_x, curr_cost
        temp = init_temp

        for i in range(max_iterations):
            perturbation = np.random.normal(0, 0.2 * temp)
            new_x = np.clip(curr_x + perturbation, 0.5, 8.0)
            new_cost = self.calculate_objective(new_x)

            if new_cost < curr_cost or np.random.random() < np.exp(-(new_cost - curr_cost) / temp):
                curr_x = new_x
                curr_cost = new_cost
                if curr_cost < best_cost:
                    best_x, best_cost = new_x, new_cost
            temp *= 0.9
        return best_x, best_cost

class Pro_UV_Optimizer:
    def __init__(self, tasks_data, num_machines, target_preference_weight, risk_magnifier=8.0):
        self.tasks_data = tasks_data
        self.num_machines = num_machines
        self.target_w = target_preference_weight
        self.risk_magnifier = risk_magnifier

    def calculate_objective(self, u, v):
        sim = FastSimulator(self.num_machines, 'pro', self.tasks_data, u=u, v=v)
        final_state = sim.run()

        m_means = final_state[1]
        m_vars = final_state[2]

        mean_of_means = np.mean(m_means)
        mean_of_vars = np.mean(m_vars)
        cv_load = np.std(m_means) / mean_of_means if mean_of_means > 0 else 0
        cv_risk = np.std(m_vars) / mean_of_vars if mean_of_vars > 0 else 0

        # exp_risk_penalty = np.exp(cv_risk * self.risk_magnifier) - 1.0
        weighted_cost = self.target_w * cv_load + (1 - self.target_w) * cv_risk
        return weighted_cost

    def optimize(self, max_iterations=30, init_temp=10.0):
        curr_u, curr_v = 3, 3
        curr_cost = self.calculate_objective(curr_u, curr_v)
        best_uv, best_cost = (curr_u, curr_v), curr_cost
        temp = init_temp

        for i in range(max_iterations):
            du = np.random.randint(-2, 3)
            dv = np.random.randint(-2, 3)
            new_u = int(np.clip(curr_u + du, 1, self.num_machines))
            new_v = int(np.clip(curr_v + dv, 1, self.num_machines))

            new_cost = self.calculate_objective(new_u, new_v)

            if new_cost < curr_cost or np.random.random() < np.exp(-(new_cost - curr_cost) / (temp + 1e-9)):
                curr_u, curr_v = new_u, new_v
                curr_cost = new_cost
                if curr_cost < best_cost:
                    best_uv, best_cost = (new_u, new_v), new_cost
            temp *= 0.9
        return best_uv, best_cost

# ==========================================
# 5. GPU 驗證與繪圖
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    print(f"\n🚀 開始 Monte Carlo 驗證 (Runs={num_runs})...")

    for name, config in strategies_config.items():
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Load Balance CDF
    ax1 = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys(), key=lambda x: 0 if 'Baseline' in x else 1)
    colors = plt.cm.viridis(np.linspace(0, 0.9, len(results)))

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8
        ax1.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax1.set_title('Load Balance CDF (Normalized)')
    ax1.set_xlim(0.9, 1.1)
    ax1.legend(fontsize='small'); ax1.grid(True, alpha=0.3)

    # Risk CDF
    ax2 = axes[1]
    base_var_key = next((k for k in results if 'Least Mu' in k), list(results.keys())[0])
    base_var_val = np.mean(results[base_var_key]['var_loads'])

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['var_loads']) / base_var_val
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8
        ax2.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax2.set_title('Risk CDF (System Variance)')
    ax2.legend(fontsize='small'); ax2.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_pareto_scatter(results):
    fig, ax = plt.subplots(figsize=(9, 6))
    points = []

    for name, data in results.items():
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

        if 'Baseline' in name:
            color = 'gray'; marker = 'X'; size = 150; w_val = -1; label = "Baseline"
        elif 'Pro' in name:
            color = 'red'; marker = '*'; size = 250; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        elif 'MOO' in name:
            color = 'blue'; marker = 'o'; size = 150; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        else:
            w_val = 0.5; color = 'green'; marker = '^'; size = 100; label = name

        points.append({'x': cv_load, 'y': cv_risk, 'c': color, 'm': marker, 's': size, 'label': label, 'w': w_val})

    for p in points:
        ax.scatter(p['x'], p['y'], c=[p['c']], s=p['s'], marker=p['m'], edgecolors='k', zorder=5)
        offset_y = 0.0005
        ax.text(p['x'], p['y'] + offset_y, p['label'], fontsize=9, ha='center', va='bottom')

    moo_points = sorted([p for p in points if 'MOO' in p['label']], key=lambda x: x['w'])
    if moo_points:
        px = [p['x'] for p in moo_points]
        py = [p['y'] for p in moo_points]
        ax.plot(px, py, 'b--', alpha=0.4, label='MOO- Frontier')

    pro_points = sorted([p for p in points if 'Pro' in p['label']], key=lambda x: x['w'])
    if pro_points:
        px = [p['x'] for p in pro_points]
        py = [p['y'] for p in pro_points]
        ax.plot(px, py, 'r-.', alpha=0.4, label='Pro Frontier')

    ax.set_title('Strategy Comparison: MOO- vs Pro (Optimized)')
    ax.set_xlabel('Load Imbalance (CV) $\\leftarrow$ Better')
    ax.set_ylabel('Risk Imbalance (CV) $\\leftarrow$ Better')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend()
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    NUM_TASKS = 3000
    NUM_MACHINES = 30
    NUM_RUNS = 1000

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 6.0

    # 建立 Task 物件 (給 MC 用)
    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    # 建立 Raw Data (給 FastSimulator SA 用，避免物件 Overhead)
    tasks_data = [(t.mean_time, t.var_time, t.actual_time) for t in task_pool]

    print("\n2. 計算 Baseline...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline System Var: {BASELINE_VAR:.2f}")

    preference_weights = [0, 0.3, 0.5, 0.7, 1.0]
    best_moo_params = {}
    best_pro_params = {}

    print("\n3. 開始雙軌優化 (MOO- vs Pro) with FastSimulator...")
    pbar = tqdm(preference_weights, desc="Optimizing Strategies")

    for w in pbar:
        # A. 優化 MOO-
        opt_moo = Threshold_X_Optimizer(tasks_data, NUM_MACHINES, w, BASELINE_VAR, risk_magnifier=8.0)
        best_x, _ = opt_moo.optimize(max_iterations=40)
        best_moo_params[w] = best_x

        # B. 優化 Pro
        opt_pro = Pro_UV_Optimizer(tasks_data, NUM_MACHINES, w, risk_magnifier=8.0)
        best_uv, _ = opt_pro.optimize(max_iterations=40)
        best_pro_params[w] = best_uv

        pbar.set_postfix({'W': w, 'x': f"{best_x:.2f}", 'uv': f"{best_uv}"})

    print("\n4. 執行評估 & 繪圖...")

    configs = {
        'Baseline (Least Mu)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}},
    }

    for w, x_val in best_moo_params.items():
        label = f'MOO- W={w} (x={x_val:.2f})'
        configs[label] = {
            'strategy': 'moo_minus',
            'params': {'var_threshold': BASELINE_VAR * x_val}
        }

    for w, (u_val, v_val) in best_pro_params.items():
        label = f'Pro W={w} (u={u_val},v={v_val})'
        configs[label] = {
            'strategy': 'pro',
            'params': {'u': u_val, 'v': v_val}
        }

    eval_results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)

    plot_cdf(eval_results, mean_times, NUM_MACHINES)
    plot_pareto_scatter(eval_results)

    print("\n✅ 優化完成！比較圖表已生成。")
```

```python
# ==============================================================================
# 完整整合代碼：MOO- vs Pro (移除模擬退火，改為手動指定參數)
# 功能：
#   1. MOO- 策略：直接指定閾值倍率 x
#   2. Pro 策略：直接指定 u, v 參數
#   3. 比較兩者在不同設定下的表現
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time', 'task_count']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0
        self.task_count = 0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time
        self.task_count += 1

# ==========================================
# 2. 模擬器 (Simulator)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='moo_minus', pre_generated_tasks=None,
                 var_threshold=float('inf'), u=3, v=3):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold
        # Pro 策略參數
        self.u = int(u)
        self.v = int(v)

    def _dispatch_moo_minus(self, task, threshold_v):
        """ MOO- 策略: Constraint + Minimize(Sum_Var) """
        task_load = task.actual_time
        current_loads = np.array([m.total_load for m in self.machines])
        candidates = []

        # 階段 1: 硬約束過濾
        for i, machine in enumerate(self.machines):
            hypothetical_loads = current_loads.copy()
            hypothetical_loads[i] += task_load
            hyp_var = np.var(hypothetical_loads)
            if hyp_var <= threshold_v:
                candidates.append(machine)

        # 階段 2: 選擇
        if not candidates:
            # Fallback
            chosen_machine = min(self.machines, key=lambda m: m.sum_mean_time)
        else:
            chosen_machine = min(candidates, key=lambda m: m.sum_var_time)
        return chosen_machine

    def _dispatch_pro(self):
        """ Pro 策略: Union(Top-u Mean, Top-v Var) -> Least Task Count """
        # 1. Mean 最小的前 u 名
        U_prime = sorted(self.machines, key=lambda m: m.sum_mean_time)[:self.u]
        # 2. Variance 最小的前 v 名
        V_prime = sorted(self.machines, key=lambda m: m.sum_var_time)[:self.v]
        # 3. 取聯集
        candidates = list(set(U_prime + V_prime))
        # 4. 選擇 task_count 最小的
        chosen_machine = min(candidates, key=lambda m: m.task_count)
        return chosen_machine

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'moo_minus':
                best_m = self._dispatch_moo_minus(task, self.var_threshold)
            elif self.strategy == 'pro':
                best_m = self._dispatch_pro()

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))
        return assignment_map

# ==========================================
# 3. GPU 驗證與繪圖
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    print(f"\n🚀 開始 Monte Carlo 驗證 (Runs={num_runs})...")

    for name, config in strategies_config.items():
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Load Balance CDF
    ax1 = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys(), key=lambda x: 0 if 'Baseline' in x else 1)
    colors = plt.cm.viridis(np.linspace(0, 0.9, len(results)))

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8

        ax1.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax1.set_title('Load Balance CDF (Normalized)')
    ax1.set_xlim(0.9, 1.1)
    ax1.legend(fontsize='small'); ax1.grid(True, alpha=0.3)

    # Risk CDF
    ax2 = axes[1]
    base_var_key = next((k for k in results if 'Least Mu' in k), list(results.keys())[0])
    base_var_val = np.mean(results[base_var_key]['var_loads'])

    for i, name in enumerate(sorted_keys):
        data = results[name]
        norm = np.sort(data['var_loads']) / base_var_val
        y = np.arange(1, len(norm)+1) / len(norm)

        if 'Baseline' in name:
            ls, lw, c, alpha = '-', 2.0, 'gray', 0.8
        elif 'Pro' in name:
            ls, lw, c, alpha = '-.', 2.5, 'red', 1.0
        elif 'MOO' in name:
            ls, lw, c, alpha = '--', 2.5, 'blue', 1.0
        else:
            ls, lw, c, alpha = ':', 2.0, colors[i], 0.8

        ax2.plot(norm, y, label=name, linewidth=lw, linestyle=ls, color=c, alpha=alpha)

    ax2.set_title('Risk CDF (System Variance)')
    ax2.legend(fontsize='small'); ax2.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_pareto_scatter(results):
    fig, ax = plt.subplots(figsize=(9, 6))
    points = []

    for name, data in results.items():
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

        if 'Baseline' in name:
            color = 'gray'; marker = 'X'; size = 150; w_val = -1; label = "Baseline"
        elif 'Pro' in name:
            color = 'red'; marker = '*'; size = 250; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        elif 'MOO' in name:
            color = 'blue'; marker = 'o'; size = 150; label = name
            w_val = float(name.split('W=')[1].split(' ')[0]) if 'W=' in name else 0.5
        else:
            w_val = 0.5; color = 'green'; marker = '^'; size = 100; label = name

        points.append({'x': cv_load, 'y': cv_risk, 'c': color, 'm': marker, 's': size, 'label': label, 'w': w_val})

    for p in points:
        ax.scatter(p['x'], p['y'], c=[p['c']], s=p['s'], marker=p['m'], edgecolors='k', zorder=5)
        offset_y = 0.0005
        ax.text(p['x'], p['y'] + offset_y, p['label'], fontsize=9, ha='center', va='bottom')

    # 畫 MOO- 的 Pareto 線
    moo_points = sorted([p for p in points if 'MOO' in p['label']], key=lambda x: x['w'])
    if moo_points:
        px = [p['x'] for p in moo_points]
        py = [p['y'] for p in moo_points]
        ax.plot(px, py, 'b--', alpha=0.4, label='MOO- Frontier')

    # 畫 Pro 的 Pareto 線
    pro_points = sorted([p for p in points if 'Pro' in p['label']], key=lambda x: x['w'])
    if pro_points:
        px = [p['x'] for p in pro_points]
        py = [p['y'] for p in pro_points]
        ax.plot(px, py, 'r-.', alpha=0.4, label='Pro Frontier')

    ax.set_title('Strategy Comparison: MOO- vs Pro (Manual Params)')
    ax.set_xlabel('Load Imbalance (CV) $\\leftarrow$ Better')
    ax.set_ylabel('Risk Imbalance (CV) $\\leftarrow$ Better')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend()
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    NUM_TASKS = 3000
    NUM_MACHINES = 30
    NUM_RUNS = 1000

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 6.0
    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print("\n2. 計算 Baseline...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline System Var: {BASELINE_VAR:.2f}")

    # =========================================================================
    # [手動參數設定區]
    # 您可以在這裡直接修改想測試的參數，無需等待 SA 搜尋
    # =========================================================================

    print("\n3. 設定手動參數...")

    # A. MOO- 策略參數: 閾值倍率 x (System_Var <= Baseline * x)
    # x 越小越嚴格 (重視 Risk)，x 越大越寬鬆 (重視 Load)
    moo_manual_params = {
        0.0: 1.05,  # 極度重視 Risk (嚴格約束)
        0.3: 1.20,
        0.5: 1.50,
        0.7: 2.00,
        1.0: 5.00   # 極度重視 Load (寬鬆約束)
    }

    # B. Pro 策略參數: (u, v)
    # u: 考慮 Mean 最小的前幾名
    # v: 考慮 Var 最小的前幾名
    # 若重視 Risk: 讓 v 小一點，u 大一點 (或根據邏輯調整)
    pro_manual_params = {
        0.0: (10, 2), # 重視 Risk (Var 候選少，從中挑 task count 小的)
        0.3: (8, 3),
        0.5: (5, 5),  # 平衡
        0.7: (3, 8),
        1.0: (2, 10)  # 重視 Load (Mean 候選少)
    }

    print("\n4. 執行評估 & 繪圖...")

    configs = {
        'Baseline (Least Mu)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}},
    }

    # 加入 MOO- 設定
    for w, x_val in moo_manual_params.items():
        label = f'MOO- W={w} (x={x_val:.2f})'
        configs[label] = {
            'strategy': 'moo_minus',
            'params': {'var_threshold': BASELINE_VAR * x_val}
        }

    # 加入 Pro 設定
    for w, (u_val, v_val) in pro_manual_params.items():
        label = f'Pro W={w} (u={u_val},v={v_val})'
        configs[label] = {
            'strategy': 'pro',
            'params': {'u': u_val, 'v': v_val}
        }

    eval_results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)

    plot_cdf(eval_results, mean_times, NUM_MACHINES)
    plot_pareto_scatter(eval_results)

    print("\n✅ 完成。")
```

```python
# ==============================================================================
# 完整整合代碼：MOO- 優化策略比較 (含方法一、二、三) + GPU 驗證
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 (用於驗證評估，不影響模擬器邏輯) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time # 模擬時使用 mean，實際評估用 Monte Carlo

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (含三種新方法)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='moo_minus', pre_generated_tasks=None,
                 var_threshold=float('inf'), sample_k=5):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold
        self.sample_k = sample_k

        # --- 全域統計量 (用於 O(1) 快速計算) ---
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    # --- 輔助：快速預測變異數 (Math Proxy) ---
    def _predict_var_fast(self, machine, task_load):
        # 使用增量公式預測加入任務後的系統變異數
        # New_Var = (New_Sum_Sq / N) - (New_Mean)^2

        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        # 預測新的平方和
        # new_sq_sum = old_sq_sum - old_load^2 + new_load^2
        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        # 預測新的總和
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)

        return predicted_var

    # --- 策略 0: 原始 MOO- (暴力掃描 + np.var) ---
    def _dispatch_original(self, task):
        task_load = task.actual_time
        # 每次都要建立 array 計算，很慢
        current_loads = np.array([m.total_load for m in self.machines])
        candidates = []

        for i, machine in enumerate(self.machines):
            hyp_loads = current_loads.copy()
            hyp_loads[i] += task_load
            if np.var(hyp_loads) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(self.machines, key=lambda m: m.total_load) # Least Load
        return min(candidates, key=lambda m: m.sum_var_time)      # Min Risk

    # --- 策略 1: Method 1 (Math Check) ---
    # 掃描全部機器，但用數學公式加速檢查
    def _dispatch_method1(self, task):
        task_load = task.actual_time
        candidates = []

        for machine in self.machines:
            # 使用 O(1) 數學預測
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)
        return min(candidates, key=lambda m: m.sum_var_time)

    # --- 策略 2: Method 2 (Sampling K + Slow Check) ---
    # 只採樣 K 台，但仍用舊的 np.var 計算 (模擬只做了採樣優化的情況)
    def _dispatch_method2(self, task):
        task_load = task.actual_time
        subset = random.sample(self.machines, min(self.num_machines, self.sample_k))

        # 為了模擬"慢"計算，這裡還是得構建全系統負載
        # 注意：這在現實中很笨，但為了對比"純採樣"與"採樣+數學"的區別，我們保留這個邏輯
        current_loads = np.array([m.total_load for m in self.machines])

        candidates = []
        for machine in subset:
            idx = machine.machine_id
            hyp_loads = current_loads.copy()
            hyp_loads[idx] += task_load
            if np.var(hyp_loads) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(subset, key=lambda m: m.total_load)
        return min(candidates, key=lambda m: m.sum_var_time)

    # --- 策略 3: Method 3 (Combined: Sampling K + Fast Math) ---
    # 最強組合：隨機 K 台 + O(1) 檢查
    def _dispatch_method3(self, task):
        task_load = task.actual_time
        subset = random.sample(self.machines, min(self.num_machines, self.sample_k))
        candidates = []

        for machine in subset:
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(subset, key=lambda m: m.total_load)
        return min(candidates, key=lambda m: m.sum_var_time)


    def run(self):
        assignment_map = []

        for task in self.tasks:
            best_m = None

            # --- 策略選擇 ---
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'original_moo':
                best_m = self._dispatch_original(task)
            elif self.strategy == 'method1':
                best_m = self._dispatch_method1(task)
            elif self.strategy == 'method2':
                best_m = self._dispatch_method2(task)
            elif self.strategy == 'method3':
                best_m = self._dispatch_method3(task)

            # --- 執行指派 ---
            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # --- 重要：更新全域統計量 (為 Fast Math 準備) ---
            # 必須在指派後立刻更新，這樣下一個任務才能拿到最新的 global 狀態
            task_load = task.actual_time
            # 更新平方和: New_Sq = Old_Sq - Old_Load^2 + New_Load^2
            # 注意：這裡的 best_m.total_load 已經在 assign_task 裡加過 task_load 了
            # 所以 Old_Load = current - task_load
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load

            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. GPU 驗證與繪圖 (不變)
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始模擬與驗證 (Runs={num_runs})...")
    print(f"{'Strategy':<25} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 55)

    for name, config in strategies_config.items():
        # --- 計時開始 ---
        start_time = time.time()

        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<25} | {elapsed:<10.4f} | {fps:<10.0f}")
        # --- 計時結束 ---

        # 準備 GPU 驗證
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            # CPU Fallback
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5)) # 3 張圖

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 3' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF'); ax.set_xlabel('Normalized Load'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    # [FIX] 這裡修正 Key Error，改抓 'Baseline (Least Mu)'
    if 'Baseline (Least Mu)' in results:
        base_ref_key = 'Baseline (Least Mu)'
    else:
        # 防呆：如果找不到，就抓第一個 Key
        base_ref_key = list(results.keys())[0]

    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        # 防呆：避免除以 0
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 3' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)'); ax.set_xlabel('Normalized Variance'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

        marker = 'o'
        size = 100
        # [FIX] 只要名字裡有 Baseline 就用 X 標記
        if 'Baseline' in name:
            marker='X'; size=150
        if 'Method 3' in name:
            size=200; marker='*'

        ax.scatter(cv_load, cv_risk, s=size, label=name, edgecolors='k', marker=marker)
        # 避免文字重疊，Baseline 往上，其他往下
        offset = 0.001 if 'Baseline' in name else -0.001
        ax.text(cv_load, cv_risk + offset, name, fontsize=9, ha='center')

    ax.set_title('Performance Trade-off (Left-Bottom is Better)')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 4. 主程式 (固定參數比較)
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    # 參數設定
    NUM_TASKS = 10000       # 任務數量
    NUM_MACHINES = 100      # 機器數量
    NUM_RUNS = 10000        # MC 驗證次數
    K_VALUE = 10            # Sampling K (採樣數)

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    # 讓變異數分布有一些極端值，凸顯 Risk 的重要性
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 8.0 # 加大風險任務的倍率

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準閥值 (基於 Least Mu 的表現)
    print("2. 計算基準閥值...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)

    # 設定閥值倍率 x
    X_MULTIPLIER = 1.1
    FIXED_THRESHOLD = BASELINE_VAR * X_MULTIPLIER
    print(f"-> Baseline (Least Mu) Var: {BASELINE_VAR:.2f}")
    print(f"-> Target Threshold (x={X_MULTIPLIER}): {FIXED_THRESHOLD:.2f}")

    # 定義要比較的策略 (加入 Least Var)
    configs = {
        # --- 兩個極端基準 ---
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        },

        # --- 原始暴力法 ---
        'Original MOO-': {
            'strategy': 'original_moo',
            'params': {'var_threshold': FIXED_THRESHOLD}
        },

        # --- 方法一：全掃描 + 快速數學 ---
        'Method 1 (Fast Math)': {
            'strategy': 'method1',
            'params': {'var_threshold': FIXED_THRESHOLD}
        },

        # --- 方法二：採樣 K=5 + 慢速數學 (模擬) ---
        f'Method 2 (Sample K={K_VALUE})': {
            'strategy': 'method2',
            'params': {'var_threshold': FIXED_THRESHOLD, 'sample_k': K_VALUE}
        },

        # --- 方法三：採樣 K=5 + 快速數學 (最強組合) ---
        f'Method 3 (Combined)': {
            'strategy': 'method3',
            'params': {'var_threshold': FIXED_THRESHOLD, 'sample_k': K_VALUE}
        },
    }

    results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
```

```python
# ==============================================================================
# 完整整合代碼：MOO- 僅使用 Fast Math (Method 1) + GPU 驗證
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 (用於驗證評估，不影響模擬器邏輯) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time # 模擬時使用 mean，實際評估用 Monte Carlo

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (專注於 Fast Math)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # --- 全域統計量 (用於 Fast Math O(1) 快速計算) ---
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    # --- 核心：快速預測變異數 (Math Proxy) ---
    def _predict_var_fast(self, machine, task_load):
        # 使用增量公式預測加入任務後的系統變異數
        # New_Var = (New_Sum_Sq / N) - (New_Mean)^2

        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        # 預測新的平方和 (O(1) 計算)
        # new_sq_sum = old_sq_sum - old_load^2 + new_load^2
        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        # 預測新的總和
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)

        return predicted_var

    # --- 策略: Method 1 (Fast Math Check) ---
    # 掃描全部機器，但使用數學公式 O(1) 加速檢查
    def _dispatch_fast_math(self, task):
        task_load = task.actual_time
        candidates = []

        for machine in self.machines:
            # 使用 O(1) 數學預測，不需構建 numpy array
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        # 如果沒有候選者滿足變異數要求，退化為 Least Load
        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)

        # 在候選者中選擇風險 (Sum Var) 最小的
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []

        for task in self.tasks:
            best_m = None

            # --- 策略選擇 ---
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'method1': # Fast Math
                best_m = self._dispatch_fast_math(task)
            else:
                # 預設行為
                best_m = self._dispatch_fast_math(task)

            # --- 執行指派 ---
            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # --- 重要：更新全域統計量 (為 Fast Math 準備) ---
            # 必須在指派後立刻更新，這樣下一個任務才能拿到最新的 global 狀態
            task_load = task.actual_time
            # 更新平方和: New_Sq = Old_Sq - Old_Load^2 + New_Load^2
            # 注意：這裡的 best_m.total_load 已經在 assign_task 裡加過 task_load 了
            # 所以 Old_Load = current - task_load
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load

            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. GPU 驗證與繪圖 (保持不變)
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始模擬與驗證 (Runs={num_runs})...")
    print(f"{'Strategy':<25} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 55)

    for name, config in strategies_config.items():
        # --- 計時開始 ---
        start_time = time.time()

        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<25} | {elapsed:<10.4f} | {fps:<10.0f}")
        # --- 計時結束 ---

        # 準備 GPU 驗證
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            # CPU Fallback
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5)) # 3 張圖

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF'); ax.set_xlabel('Normalized Load'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    if 'Baseline (Least Mu)' in results:
        base_ref_key = 'Baseline (Least Mu)'
    else:
        base_ref_key = list(results.keys())[0]

    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)'); ax.set_xlabel('Normalized Variance'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

        marker = 'o'
        size = 100
        if 'Baseline' in name:
            marker='X'; size=150
        if 'Method 1' in name:
            size=200; marker='*'

        ax.scatter(cv_load, cv_risk, s=size, label=name, edgecolors='k', marker=marker)
        offset = 0.001 if 'Baseline' in name else -0.001
        ax.text(cv_load, cv_risk + offset, name, fontsize=9, ha='center')

    ax.set_title('Performance Trade-off (Left-Bottom is Better)')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 4. 主程式 (僅執行 Fast Math)
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    # 參數設定
    NUM_TASKS = 10000       # 任務數量
    NUM_MACHINES = 100      # 機器數量
    NUM_RUNS = 10000        # MC 驗證次數

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 8.0

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準閥值 (基於 Least Mu 的表現)
    print("2. 計算基準閥值...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)

    # 設定閥值倍率 x
    X_MULTIPLIER = 0
    FIXED_THRESHOLD = BASELINE_VAR * X_MULTIPLIER
    print(f"-> Baseline (Least Mu) Var: {BASELINE_VAR:.2f}")
    print(f"-> Target Threshold (x={X_MULTIPLIER}): {FIXED_THRESHOLD:.2f}")

    # 定義要比較的策略 (僅保留 Fast Math 與基準線)
    configs = {
        # --- 基準線 (保留作為比較參考) ---
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        },

        # --- 唯一主角：方法一 (Fast Math) ---
        'Method 1 (Fast Math)': {
            'strategy': 'method1',
            'params': {'var_threshold': FIXED_THRESHOLD}
        },
    }

    results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
```

```python
# ==============================================================================
# 完整整合代碼：MOO- Fast Math + SA 參數優化 (尋找最佳 X_MULTIPLIER)
# ==============================================================================

import numpy as np
import random
import matplotlib.pyplot as plt
import time
import math
from tqdm import tqdm

# --- GPU 加速檢查 (僅用於最終驗證，不影響 SA 過程) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (Fast Math)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # 全域統計量 (Fast Math 用)
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    def _predict_var_fast(self, machine, task_load):
        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)
        return predicted_var

    def _dispatch_fast_math(self, task):
        task_load = task.actual_time
        candidates = []

        for machine in self.machines:
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            else:
                best_m = self._dispatch_fast_math(task)

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # 更新全域狀態
            task_load = task.actual_time
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load
            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. 模擬退火優化 (SA Optimization)
# ==========================================

def calculate_cost(x_mult, base_var, tasks, num_machines):
    """
    計算成本函數: Cost = CV(Load) + CV(Risk)
    注意：這裡使用 Simulator 內部的 sum_var_time 作為 Risk 的 Proxy，
    不執行昂貴的 MC 模擬，以確保 SA 速度。
    """
    threshold = base_var * x_mult
    sim = Simulator(num_machines, strategy='method1',
                    pre_generated_tasks=tasks, var_threshold=threshold)
    sim.run()

    # 提取統計數據
    loads = np.array([m.total_load for m in sim.machines])
    risks = np.array([m.sum_var_time for m in sim.machines])

    # 計算 Coefficient of Variation (CV) = Std / Mean
    # 加一個極小值 eps 防止除以零
    eps = 1e-9
    cv_load = np.std(loads) / (np.mean(loads) + eps)
    cv_risk = np.std(risks) / (np.mean(risks) + eps)

    return cv_load + cv_risk, cv_load, cv_risk

def simulated_annealing(base_var, tasks, num_machines):
    print("\n🔥 啟動模擬退火 (Simulated Annealing) 尋找最佳 X...")

    # --- SA 參數設定 ---
    T_initial = 1.0       # 初始溫度
    T_min = 0.001         # 終止溫度
    alpha = 0.95          # 降溫係數 (Cooling Rate)
    iter_per_temp = 20    # 每個溫度下的迭代次數

    # --- 狀態初始化 ---
    current_x = 1.1       # 初始猜測值
    current_cost, _, _ = calculate_cost(current_x, base_var, tasks, num_machines)

    best_x = current_x
    best_cost = current_cost

    history_cost = []
    history_x = []

    T = T_initial

    # 進度條 (總迭代數估算)
    total_steps = int(math.log(T_min / T_initial) / math.log(alpha)) * iter_per_temp
    pbar = tqdm(total=total_steps, desc="SA Optimization")

    while T > T_min:
        for _ in range(iter_per_temp):
            # 1. 產生新鄰居 (在目前 X 附近隨機擾動)
            # 隨著溫度降低，擾動範圍變小 (自適應步長)
            step_size = max(0.05, 0.5 * (T / T_initial))
            neighbor_x = current_x + random.uniform(-step_size, step_size)

            # 邊界限制 (X 必須 > 0，且通常不需要大於 5.0)
            neighbor_x = max(0.1, min(5.0, neighbor_x))

            # 2. 計算新成本
            neighbor_cost, _, _ = calculate_cost(neighbor_x, base_var, tasks, num_machines)

            # 3. 接受準則 (Metropolis Criterion)
            delta = neighbor_cost - current_cost

            # 如果新成本更低 (delta < 0)，必定接受
            # 如果新成本更高，以機率 P = exp(-delta / T) 接受
            if delta < 0 or random.random() < math.exp(-delta / T):
                current_x = neighbor_x
                current_cost = neighbor_cost

                # 更新全域最佳解
                if current_cost < best_cost:
                    best_cost = current_cost
                    best_x = current_x

            history_cost.append(current_cost)
            history_x.append(current_x)
            pbar.update(1)

        # 4. 降溫
        T *= alpha

    pbar.close()

    # 最終最佳解的詳細數據
    final_cost, final_cv_load, final_cv_risk = calculate_cost(best_x, base_var, tasks, num_machines)

    print(f"✅ SA 完成! 最佳 X_MULTIPLIER: {best_x:.4f}")
    print(f"   Min Cost: {best_cost:.5f} (CV_Load: {final_cv_load:.4f} + CV_Risk: {final_cv_risk:.4f})")

    # 繪製收斂過程
    plt.figure(figsize=(10, 4))
    plt.plot(history_cost, label='Cost (CV_L + CV_R)', alpha=0.7)
    plt.axhline(y=best_cost, color='r', linestyle='--', label=f'Best: {best_cost:.4f}')
    plt.title(f'Simulated Annealing Convergence (Best X={best_x:.3f})')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    return best_x

# ==========================================
# 4. GPU 驗證與繪圖 (保持不變)
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始最終驗證 (MC Runs={num_runs})...")
    print(f"{'Strategy':<30} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 60)

    for name, config in strategies_config.items():
        start_time = time.time()
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()
        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<30} | {elapsed:<10.4f} | {fps:<10.0f}")

        # GPU 驗證邏輯
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines, best_x_val):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'SA Optimized' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    base_ref_key = 'Baseline (Least Mu)'
    if base_ref_key not in results: base_ref_key = list(results.keys())[0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'SA Optimized' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        total_cost = cv_load + cv_risk

        marker = 'o'; size = 100
        if 'Baseline' in name: marker='X'; size=150
        if 'SA Optimized' in name: size=250; marker='*'

        ax.scatter(cv_load, cv_risk, s=size, label=f"{name}\n(Cost={total_cost:.3f})", edgecolors='k', marker=marker)

    ax.set_title(f'Performance Trade-off (Best X={best_x_val:.2f})')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.legend(fontsize='small')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    # 參數設定
    NUM_TASKS = 10000
    NUM_MACHINES = 100
    NUM_RUNS = 10000

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 8.0

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準 Variance (作為 X_MULTIPLIER 的單位)
    print("2. 計算基準閥值 (Least Mu)...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline Var: {BASELINE_VAR:.2f}")

    # 3. 執行模擬退火 (SA) 尋找最佳 X
    # 目標函數: Cost = CV_Load + CV_Risk
    best_x = simulated_annealing(BASELINE_VAR, task_pool, NUM_MACHINES)

    # 4. 進行最終比較 (Baseline vs. SA Optimized)
    best_threshold = BASELINE_VAR * best_x

    configs = {
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        },
        # 使用 SA 找到的最佳參數
        f'Method 1 (SA Optimized X={best_x:.2f})': {
            'strategy': 'method1',
            'params': {'var_threshold': best_threshold}
        },
        # 對照組：使用一個固定的隨機 X (例如 1.1)
        f'Method 1 (Fixed X=1.1)': {
            'strategy': 'method1',
            'params': {'var_threshold': BASELINE_VAR * 1.1}
        },
    }

    results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES, best_x)
```

```python
# ==============================================================================
# 完整整合代碼：MOO - Fast Math 極端值測試 (Th=0 vs Th=Inf)
# ==============================================================================

import numpy as np
import collections
import random
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# --- GPU 加速檢查 (用於驗證評估，不影響模擬器邏輯) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time # 模擬時使用 mean，實際評估用 Monte Carlo

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (專注於 Fast Math)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # --- 全域統計量 (用於 Fast Math O(1) 快速計算) ---
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    # --- 核心：快速預測變異數 (Math Proxy) ---
    def _predict_var_fast(self, machine, task_load):
        # 使用增量公式預測加入任務後的系統變異數
        # New_Var = (New_Sum_Sq / N) - (New_Mean)^2

        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        # 預測新的平方和 (O(1) 計算)
        # new_sq_sum = old_sq_sum - old_load^2 + new_load^2
        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        # 預測新的總和
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)

        return predicted_var

    # --- 策略: Method 1 (Fast Math Check) ---
    # 掃描全部機器，但使用數學公式 O(1) 加速檢查
    def _dispatch_fast_math(self, task):
        task_load = task.actual_time
        candidates = []

        for machine in self.machines:
            # 使用 O(1) 數學預測
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        # 如果沒有候選者滿足變異數要求，退化為 Least Load (類似 Least Mu)
        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)

        # 在候選者中選擇風險 (Sum Var) 最小的
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []

        for task in self.tasks:
            best_m = None

            # --- 策略選擇 ---
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'method1': # Fast Math
                best_m = self._dispatch_fast_math(task)
            else:
                best_m = self._dispatch_fast_math(task)

            # --- 執行指派 ---
            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # --- 重要：更新全域統計量 (為 Fast Math 準備) ---
            task_load = task.actual_time
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load

            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. GPU 驗證與繪圖
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始模擬與驗證 (Runs={num_runs})...")
    print(f"{'Strategy':<30} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 60)

    for name, config in strategies_config.items():
        # --- 計時開始 ---
        start_time = time.time()

        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<30} | {elapsed:<10.4f} | {fps:<10.0f}")

        # 準備 GPU 驗證
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            # CPU Fallback
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 2, figsize=(18, 5)) # 3 張圖

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)

        # 樣式調整
        lw = 2.5 if 'Method 1' in name else 1.5
        ls = '--' if 'Baseline' in name else '-'

        ax.plot(norm, y, label=name, lw=lw, linestyle=ls)
    ax.set_title('Load Balance CDF'); ax.set_xlabel('Normalized Load'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    # 找一個 Baseline 當分母
    base_ref_key = [k for k in results.keys() if 'Least Mu' in k][0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)

        lw = 2.5 if 'Method 1' in name else 1.5
        ls = '--' if 'Baseline' in name else '-'

        ax.plot(norm, y, label=name, lw=lw, linestyle=ls)
    ax.set_title('Risk CDF (Normalized)'); ax.set_xlabel('Normalized Variance'); ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # # 3. Pareto Scatter
    # ax = axes[2]
    # colors = plt.cm.get_cmap('tab10', len(results))

    # for i, name in enumerate(sorted_keys):
    #     data = results[name]
    #     cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
    #     cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])

    #     marker = 'o'
    #     size = 100
    #     if 'Baseline' in name:
    #         marker='X'; size=150
    #     if 'Method 1' in name:
    #         size=200; marker='*'

    #     ax.scatter(cv_load, cv_risk, s=size, label=name, edgecolors='k', marker=marker, alpha=0.8)

    #     # 文字標籤偏移，避免重疊
    #     offset_y = 0.002 if i % 2 == 0 else -0.002
    #     ax.text(cv_load, cv_risk + offset_y, name, fontsize=9, ha='center', fontweight='bold')

    # ax.set_title('Performance Trade-off (Left-Bottom is Better)')
    # ax.set_xlabel('Load Imbalance (CV)')
    # ax.set_ylabel('Risk Imbalance (CV)')
    # ax.grid(True, linestyle='--', alpha=0.5)

    # plt.tight_layout()
    # plt.show()

# ==========================================
# 4. 主程式 (修改：展示極端情況)
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    # 參數設定
    NUM_TASKS = 10000       # 任務數量
    NUM_MACHINES = 100      # 機器數量
    NUM_RUNS = 10000        # MC 驗證次數

    print("1. 生成數據...")
    mean_times = np.random.uniform(10, 100, NUM_TASKS)
    var_times = np.random.gamma(shape=2.0, scale=15.0, size=NUM_TASKS)
    risky_idx = np.random.choice(NUM_TASKS, int(NUM_TASKS*0.05), replace=False)
    var_times[risky_idx] *= 8.0

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 定義要比較的策略：加入 Th=0 與 Th=Inf
    configs = {
        # --- 基準線 ---
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        },

        # --- Method 1 極端情況 ---
        'Method 1 (Th=0)': {
            'strategy': 'method1',
            # 閾值為 0 -> 強制失敗 -> 強制 Fallback -> 等同於 Load Balancing
            'params': {'var_threshold': 0.0}
        },
        'Method 1 (Th=Inf)': {
            'strategy': 'method1',
            # 閾值無限大 -> 永遠通過 -> 選擇最小風險 -> 等同於 Risk Minimization
            'params': {'var_threshold': float('inf')}
        },
    }

    results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
```

```python
# ==============================================================================
# 完整整合代碼：MOO- Fast Math + SA 參數優化 (尋找最佳 X_MULTIPLIER)
# ==============================================================================

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import time
import math
from tqdm import tqdm

# --- GPU 加速檢查 (僅用於最終驗證，不影響 SA 過程) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (Fast Math)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # 全域統計量 (Fast Math 用)
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    def _predict_var_fast(self, machine, task_load):
        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)
        return predicted_var

    def _dispatch_fast_math(self, task):
        task_load = task.actual_time
        candidates = []

        for machine in self.machines:
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            else:
                best_m = self._dispatch_fast_math(task)

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # 更新全域狀態
            task_load = task.actual_time
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load
            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. 模擬退火優化 (SA Optimization)
# ==========================================

def calculate_cost(x_mult, base_var, tasks, num_machines):
    """
    計算成本函數: Cost = CV(Load) + CV(Risk)
    注意：這裡使用 Simulator 內部的 sum_var_time 作為 Risk 的 Proxy，
    不執行昂貴的 MC 模擬，以確保 SA 速度。
    """
    threshold = base_var * x_mult
    sim = Simulator(num_machines, strategy='method1',
                    pre_generated_tasks=tasks, var_threshold=threshold)
    sim.run()

    # 提取統計數據
    loads = np.array([m.total_load for m in sim.machines])
    risks = np.array([m.sum_var_time for m in sim.machines])

    # 計算 Coefficient of Variation (CV) = Std / Mean
    # 加一個極小值 eps 防止除以零
    eps = 1e-9
    cv_load = np.std(loads) / (np.mean(loads) + eps)
    cv_risk = np.std(risks) / (np.mean(risks) + eps)

    return cv_load + cv_risk, cv_load, cv_risk

def simulated_annealing(base_var, tasks, num_machines):
    print("\n🔥 啟動模擬退火 (Simulated Annealing) 尋找最佳 X...")

    # --- SA 參數設定 ---
    T_initial = 1.0       # 初始溫度
    T_min = 0.001         # 終止溫度
    alpha = 0.95          # 降溫係數 (Cooling Rate)
    iter_per_temp = 20    # 每個溫度下的迭代次數

    # --- 狀態初始化 ---
    current_x = 1.1       # 初始猜測值
    current_cost, _, _ = calculate_cost(current_x, base_var, tasks, num_machines)

    best_x = current_x
    best_cost = current_cost

    history_cost = []
    history_x = []

    T = T_initial

    # 進度條 (總迭代數估算)
    total_steps = int(math.log(T_min / T_initial) / math.log(alpha)) * iter_per_temp
    pbar = tqdm(total=total_steps, desc="SA Optimization")

    while T > T_min:
        for _ in range(iter_per_temp):
            # 1. 產生新鄰居 (在目前 X 附近隨機擾動)
            # 隨著溫度降低，擾動範圍變小 (自適應步長)
            step_size = max(0.05, 0.5 * (T / T_initial))
            neighbor_x = current_x + random.uniform(-step_size, step_size)

            # 邊界限制 (X 必須 > 0，且通常不需要大於 5.0)
            neighbor_x = max(0.1, min(5.0, neighbor_x))

            # 2. 計算新成本
            neighbor_cost, _, _ = calculate_cost(neighbor_x, base_var, tasks, num_machines)

            # 3. 接受準則 (Metropolis Criterion)
            delta = neighbor_cost - current_cost

            # 如果新成本更低 (delta < 0)，必定接受
            # 如果新成本更高，以機率 P = exp(-delta / T) 接受
            if delta < 0 or random.random() < math.exp(-delta / T):
                current_x = neighbor_x
                current_cost = neighbor_cost

                # 更新全域最佳解
                if current_cost < best_cost:
                    best_cost = current_cost
                    best_x = current_x

            history_cost.append(current_cost)
            history_x.append(current_x)
            pbar.update(1)

        # 4. 降溫
        T *= alpha

    pbar.close()

    # 最終最佳解的詳細數據
    final_cost, final_cv_load, final_cv_risk = calculate_cost(best_x, base_var, tasks, num_machines)

    print(f"✅ SA 完成! 最佳 X_MULTIPLIER: {best_x:.4f}")
    print(f"   Min Cost: {best_cost:.5f} (CV_Load: {final_cv_load:.4f} + CV_Risk: {final_cv_risk:.4f})")

    # 繪製收斂過程
    plt.figure(figsize=(10, 4))
    plt.plot(history_cost, label='Cost (CV_L + CV_R)', alpha=0.7)
    plt.axhline(y=best_cost, color='r', linestyle='--', label=f'Best: {best_cost:.4f}')
    plt.title(f'Simulated Annealing Convergence (Best X={best_x:.3f})')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    return best_x

# ==========================================
# 4. GPU 驗證與繪圖 (保持不變)
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始最終驗證 (MC Runs={num_runs})...")
    print(f"{'Strategy':<30} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 60)

    for name, config in strategies_config.items():
        start_time = time.time()
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()
        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<30} | {elapsed:<10.4f} | {fps:<10.0f}")

        # GPU 驗證邏輯
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(1e-6, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)
                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines, best_x_val):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'SA Optimized' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    base_ref_key = 'Baseline (Least Mu)'
    if base_ref_key not in results: base_ref_key = list(results.keys())[0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'SA Optimized' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        total_cost = cv_load + cv_risk

        marker = 'o'; size = 100
        if 'Baseline' in name: marker='X'; size=150
        if 'SA Optimized' in name: size=250; marker='*'

        ax.scatter(cv_load, cv_risk, s=size, label=f"{name}\n(Cost={total_cost:.3f})", edgecolors='k', marker=marker)

    ax.set_title(f'Performance Trade-off (Best X={best_x_val:.2f})')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.legend(fontsize='small')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42); random.seed(42)

    # 參數設定
    NUM_MACHINES = 100
    NUM_RUNS = 10000

    print("1. 讀取 google.csv 數據...")
    try:
        df = pd.read_csv('google.csv')
        print(f"成功讀取 google.csv，共 {len(df)} 筆資料")
    except FileNotFoundError:
        print("錯誤: 找不到 google.csv，請確認檔案位置")
        exit()

    # 使用 google.csv 中的數據
    # 假設 CSV 欄位名稱為 'average_usage_mean' 和 'average_usage_variance'
    if 'average_usage_mean' not in df.columns or 'average_usage_variance' not in df.columns:
        print("錯誤: CSV 檔案缺少必要的欄位 (average_usage_mean, average_usage_variance)")
        # 嘗試使用欄位索引 (如果欄位名稱不匹配)
        print("嘗試使用第 2 和第 3 欄位作為 mean 和 variance...")
        mean_times = df.iloc[:, 1].values
        var_times = df.iloc[:, 2].values
    else:
        mean_times = df['average_usage_mean'].values
        var_times = df['average_usage_variance'].values

    # 確保變異數為正 (有些數據可能是 0，給予一個極小值)
    var_times = np.maximum(var_times, 1e-9)

    # 為了避免數值過小導致計算問題，可以考慮是否需要縮放 (Scale)
    # 例如：如果 mean 都是 0.00x，可能需要放大，但這裡先保持原樣，
    # 因為 Simulator 邏輯是通用的。

    NUM_TASKS = len(mean_times)

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準 Variance (作為 X_MULTIPLIER 的單位)
    print("2. 計算基準閥值 (Least Mu)...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline Var: {BASELINE_VAR:.2f}")

    # 3. 執行模擬退火 (SA) 尋找最佳 X
    # 目標函數: Cost = CV_Load + CV_Risk
    best_x = simulated_annealing(BASELINE_VAR, task_pool, NUM_MACHINES)

    # 4. 進行最終比較 (Baseline vs. SA Optimized)
    best_threshold = BASELINE_VAR * best_x

    configs = {
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        },
        # 使用 SA 找到的最佳參數
        f'Method 1 (SA Optimized X={best_x:.2f})': {
            'strategy': 'method1',
            'params': {'var_threshold': best_threshold}
        },
        # 對照組：使用一個固定的隨機 X (例如 1.1)
        f'Method 1 (Fixed X=1.1)': {
            'strategy': 'method1',
            'params': {'var_threshold': BASELINE_VAR * 1.1}
        },
    }

    results = run_evaluation_comparison(mean_times, var_times, configs, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES, best_x)
```

```python
# ==============================================================================
# 完整整合代碼：MOO- Fast Math + 手動指定 X 參數 (無 SA)
# ==============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import math

# --- GPU 加速檢查 (僅用於最終驗證) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (Fast Math)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # 全域統計量 (Fast Math 用)
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

    def _predict_var_fast(self, machine, task_load):
        n = self.num_machines
        old_load = machine.total_load
        new_load = old_load + task_load

        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)
        return predicted_var

    def _dispatch_fast_math(self, task):
        task_load = task.actual_time
        candidates = []

        # 這裡找出所有 "分配後 Variance 不超過閾值" 的機器
        for machine in self.machines:
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        # 如果沒有機器符合要求，退化為 Least Load (Greedy)
        if not candidates:
            return min(self.machines, key=lambda m: m.total_load)

        # 如果有候選者，從中選 Risk (Sum Var) 最小的
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            else:
                best_m = self._dispatch_fast_math(task)

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # 更新全域狀態
            task_load = task.actual_time
            old_load = best_m.total_load - task_load
            new_load = best_m.total_load
            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. GPU 驗證與繪圖 (核心邏輯不變)
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}
    # 準備 GPU 數據
    try:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))
        use_gpu = True
    except:
        means_gpu = np.array(mean_times, dtype=np.float32)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float32))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    print(f"\n🚀 開始最終驗證 (MC Runs={num_runs})...")
    print(f"{'Strategy':<35} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 65)

    for name, config in strategies_config.items():
        start_time = time.time()
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()
        elapsed = time.time() - start_time
        fps = len(task_pool) / elapsed if elapsed > 0 else 0
        print(f"{name:<35} | {elapsed:<10.4f} | {fps:<10.0f}")

        # --- GPU MC 模擬驗證 ---
        if use_gpu:
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                # 產生隨機負載 (常態分佈)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float32)
                batch_rand = cp.maximum(1e-6, batch_rand)

                # 矩陣乘法計算機器負載
                batch_loads = cp.dot(batch_rand, assign_matrix)

                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)

                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            # CPU 簡易模式 (僅累加 Mean/Var，非 MC)
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}
    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    base_ref_key = 'Baseline (Least Mu)'
    if base_ref_key not in results: base_ref_key = list(results.keys())[0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        # 計算變異係數 (Coefficient of Variation)
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        total_cost = cv_load + cv_risk

        marker = 'o'; size = 100
        if 'Baseline' in name: marker='X'; size=150
        if 'Method 1' in name: size=200; marker='D'

        ax.scatter(cv_load, cv_risk, s=size, label=f"{name}\n(Cost={total_cost:.3f})", edgecolors='k', marker=marker)

    ax.set_title('Performance Trade-off')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.legend(fontsize='x-small', loc='upper right')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 4. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42)

    # 參數設定
    NUM_MACHINES = 96
    NUM_RUNS = 10000

    # 在這裡設定你想測試的 X 倍率 (手動指定，取代 SA)
    # 例如：[1.0, 1.2, 1.5, 2.0]
    TEST_MULTIPLIERS = [10000]

    print("1. 讀取 google.csv 數據...")
    try:
        df = pd.read_csv('google.csv')
        print(f"成功讀取 google.csv，共 {len(df)} 筆資料")
    except FileNotFoundError:
        print("錯誤: 找不到 google.csv，請確認檔案位置")
        # 產生假資料以便測試
        print("⚠️ 產生隨機假資料進行演示...")
        df = pd.DataFrame({
            'average_usage_mean': np.random.uniform(0.5, 5.0, 5000),
            'average_usage_variance': np.random.uniform(0.1, 2.0, 5000)
        })

    if 'average_usage_mean' not in df.columns:
        mean_times = df.iloc[:, 1].values
        var_times = df.iloc[:, 2].values
    else:
        mean_times = df['average_usage_mean'].values
        var_times = df['average_usage_variance'].values


    # 讀取後放大 1,000,000 倍 (1e6)
    mean_times = df['average_usage_mean'].values * 1e6
    var_times = df['average_usage_variance'].values * 1e12 # 或是 1e12，視您對 Risk 的定義

    var_times = np.maximum(var_times, 1e-20)
    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準 Variance
    print("2. 計算基準閥值 (Least Mu)...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    base_loads = [m.total_load for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline Var: {BASELINE_VAR:.20f}")

    # 3. 建立測試配置 (Configs)
    strategies_config = {
        'Baseline (Least Mu)': {
            'strategy': 'least_mu',
            'params': {}
        },
        'Baseline (Least Var)': {
            'strategy': 'least_var',
            'params': {}
        }
    }

    # 自動加入 TEST_MULTIPLIERS 中的設定
    for x in TEST_MULTIPLIERS:
        threshold = BASELINE_VAR * x
        strategies_config[f'Method 1 (X={x:.1f})'] = {
            'strategy': 'method1',
            'params': {'var_threshold': threshold}
        }

    # 4. 執行與比較
    results = run_evaluation_comparison(mean_times, var_times, strategies_config, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
```

```python
# ==============================================================================
# 完整整合代碼：MOO- Fast Math + Fallback Counter + 96000 Tasks Limit (Refactored)
# ==============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import math

# --- GPU 加速檢查 ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")

# ==========================================
# 1. 資料結構 (已移除 actual_time 與 total_load)
# ==========================================

class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time']
    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        # actual_time 已移除，直接使用 mean_time

class Machine:
    __slots__ = ['machine_id', 'sum_mean_time', 'sum_var_time']
    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        # total_load 已移除，直接使用 sum_mean_time
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0

    def assign_task(self, task):
        # total_load 的累加動作合併至 sum_mean_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time

# ==========================================
# 2. 模擬器 (邏輯更新)
# ==========================================

class Simulator:
    def __init__(self, num_machines, strategy='method1', pre_generated_tasks=None,
                 var_threshold=float('inf')):

        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.tasks = pre_generated_tasks
        self.strategy = strategy
        self.var_threshold = var_threshold

        # 全域統計量 (這裡的 Load 指的是 Mean Time 的總和)
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0

        # === Fallback 計數器 ===
        self.fallback_count = 0
        self.sigma_dispatch_count = 0

    def _predict_var_fast(self, machine, task_load):
        n = self.num_machines
        # 修改：原本使用 total_load，現在使用 sum_mean_time
        old_load = machine.sum_mean_time
        new_load = old_load + task_load

        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)
        return predicted_var

    def _dispatch_fast_math(self, task):
        task_load = task.mean_time

        # 1. 取得所有機器的當前負載 (sum_mean_time)
        loads = [m.sum_mean_time for m in self.machines]
        n = len(loads)

        if n > 1:
            avg_load = np.mean(loads)
            var_load = np.var(loads)
            std_load=var_load**0.5
            threshold = avg_load - (2 * std_load)

            # 5. 找出在分佈中「小於 3 倍標準差」的節點
            sigma_candidates = [m for m in self.machines if m.sum_mean_time < threshold]

            # 6. 若有候選人符合條件 (代表系統中有非極端異常的機器)
            if sigma_candidates:
                # 依照先前指示：直接 return 所有機器 (self.machines) 中 mean_time 最小者
                self.sigma_dispatch_count += 1
                return min(self.machines, key=lambda m: m.sum_mean_time)

        # (若機器數量 <=1 或是極端狀況沒人符合，則繼續往下執行)

        # === 原有階段：變異數閾值過濾 ===
        candidates = []
        for machine in self.machines:
            if self._predict_var_fast(machine, task_load) <= self.var_threshold:
                candidates.append(machine)

        if not candidates:
            # === 觸發 Fallback (退化為 Least Mu) ===
            self.fallback_count += 1
            return min(self.machines, key=lambda m: m.sum_mean_time)

        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        assignment_map = []
        for task in self.tasks:
            best_m = None
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            else:
                best_m = self._dispatch_fast_math(task)

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

            # 更新全域統計量
            task_load = task.mean_time
            # 由於 assign_task 已經執行，新的 sum_mean_time 就是 new_load
            new_load = best_m.sum_mean_time
            old_load = new_load - task_load

            self.global_sum_load += task_load
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)

        return assignment_map

# ==========================================
# 3. 驗證與輸出
# ==========================================

def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}

    # --- GPU 數據準備 ---
    try:
        import cupy as cp
        # 嘗試使用 GPU 記憶體池以優化效能
        mempool = cp.get_default_memory_pool()
        means_gpu = cp.array(mean_times, dtype=cp.float64)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float64))
        use_gpu = True
    except ImportError:
        import numpy as np
        means_gpu = np.array(mean_times, dtype=np.float64)
        stds_gpu = np.sqrt(np.array(var_times, dtype=np.float64))
        use_gpu = False

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    total_tasks = len(task_pool)

    print(f"\n🚀 開始最終驗證 (Tasks={total_tasks}, MC Runs={num_runs})...")
    # [修改] 新增 3-Sigma 欄位
    print(f"{'Strategy':<25} | {'Time(s)':<8} | {'FPS':<7} | {'Fallback (Rate)':<20} | {'3-Sigma (Rate)':<20}")
    print("-" * 90)

    for name, config in strategies_config.items():
        start_time = time.time()

        # 初始化模擬器
        sim = Simulator(num_machines, strategy=config['strategy'],
                        pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = total_tasks / elapsed if elapsed > 0 else 0

        # === [修改] 格式化計數器輸出 ===
        # 使用 getattr 安全讀取，若演算法無此屬性則預設為 0
        fallback_cnt = getattr(sim, 'fallback_count', 0)
        sigma_cnt = getattr(sim, 'sigma_dispatch_count', 0)

        # 僅針對有數據的策略顯示 (通常是 method1)
        if fallback_cnt > 0 or sigma_cnt > 0 or config['strategy'] == 'method1':
            fb_rate = (fallback_cnt / total_tasks) * 100
            fb_str = f"{fallback_cnt} ({fb_rate:.1f}%)"

            sig_rate = (sigma_cnt / total_tasks) * 100
            sig_str = f"{sigma_cnt} ({sig_rate:.1f}%)"
        else:
            fb_str = "N/A"
            sig_str = "N/A"

        print(f"{name:<25} | {elapsed:<8.4f} | {fps:<7.0f} | {fb_str:<20} | {sig_str:<20}")

        # --- Monte Carlo 驗證 (GPU/CPU) ---
        if use_gpu:
            # 建立分配矩陣 (Task x Machine)
            assign_matrix = cp.zeros((len(mean_times), num_machines), dtype=cp.float64)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float64)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float64)

            # 分批執行以避免顯存溢出
            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                # 生成常態分佈隨機數
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, len(mean_times))).astype(cp.float64)
                batch_rand = cp.maximum(1e-20, batch_rand) # 避免負值

                # 矩陣乘法計算負載
                batch_loads = cp.dot(batch_rand, assign_matrix)

                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)

                # 釋放記憶體
                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)

            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            # CPU Fallback (簡單加總平均值與變異數，僅供參考，不進行 MC 模擬以節省時間)
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}

    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Load Balance CDF')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    base_ref_key = 'Baseline (Least Mu)'
    if base_ref_key not in results: base_ref_key = list(results.keys())[0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])

    for name in sorted_keys:
        data = results[name]
        denominator = base_var_val if base_var_val > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)
        lw = 2.5 if 'Method 1' in name else 1.5
        ax.plot(norm, y, label=name, lw=lw)
    ax.set_title('Risk CDF (Normalized)')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        total_cost = cv_load + cv_risk

        marker = 'o'; size = 100
        if 'Baseline' in name: marker='X'; size=150
        if 'Method 1' in name: size=200; marker='D'

        ax.scatter(cv_load, cv_risk, s=size, label=f"{name}\n(Cost={total_cost:.3f})", edgecolors='k', marker=marker)

    ax.set_title('Performance Trade-off')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.legend(fontsize='x-small', loc='upper right')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 4. 主程式
# ==========================================

if __name__ == "__main__":
    np.random.seed(42)

    # 參數設定
    NUM_MACHINES = 100
    NUM_RUNS = 10000
    TEST_MULTIPLIERS = [10000]

    print("1. 讀取 google.csv 數據...")
    try:
        df = pd.read_csv('google.csv')
        print(f"成功讀取 google.csv，原始資料共 {len(df)} 筆")
    except FileNotFoundError:
        print("錯誤: 找不到 google.csv")
        exit()

    if 'average_usage_mean' not in df.columns:
        mean_times = df.iloc[:, 1].values
        var_times = df.iloc[:, 2].values
    else:
        mean_times = df['average_usage_mean'].values
        var_times = df['average_usage_variance'].values

    # === 數據縮放 ===
    print("⚖️ 執行數據縮放 (Scaling x 1e6)...")
    # mean_times = mean_times * 1e6
    # var_times = var_times * 1e5
    # var_times = np.maximum(var_times, 1e-20)

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]

    # 2. 計算基準
    print("2. 計算基準閥值 (Least Mu)...")
    base_sim = Simulator(NUM_MACHINES, strategy='least_mu', pre_generated_tasks=task_pool)
    base_sim.run()
    # 修改：從 total_load 改為 sum_mean_time
    base_loads = [m.sum_mean_time for m in base_sim.machines]
    BASELINE_VAR = np.var(base_loads)
    print(f"-> Baseline Var: {BASELINE_VAR:.4f}")

    # 3. 建立測試配置
    strategies_config = {
        'Baseline (Least Mu)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}}
    }
    for x in TEST_MULTIPLIERS:
        threshold = BASELINE_VAR * x
        strategies_config[f'Method 1 (X={x:.1f})'] = {
            'strategy': 'method1',
            'params': {'var_threshold': threshold}
        }

    # 4. 執行與比較
    results = run_evaluation_comparison(mean_times, var_times, strategies_config, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
```

```python
# ==============================================================================
# 完整整合代碼：Baseline vs 論文雙目標演算法 (Paper Migration) + GPU MC 驗證
# 數據源：/content/google.csv
# ==============================================================================

import random
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- GPU 加速檢查 (用於 MC 驗證) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
    USE_GPU = True
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")
    USE_GPU = False

# ==========================================
# 1. 資料結構
# ==========================================
class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']

    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time', 'tasks']

    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0
        self.tasks = []

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time
        self.tasks.append(task)

    def remove_task(self, task):
        self.total_load -= task.actual_time
        self.sum_mean_time -= task.mean_time
        self.sum_var_time -= task.var_time
        self.tasks.remove(task)

# ==========================================
# 2. 模擬器 (包含論文去中心化演算法)
# ==========================================
class Simulator:
    def __init__(self, num_machines, strategy='least_mu', pre_generated_tasks=None, **kwargs):
        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.strategy = strategy
        self.tasks = pre_generated_tasks if pre_generated_tasks else []

        self.alpha = kwargs.get('alpha', 1.0)
        self.beta = kwargs.get('beta', 1.0)

    def _run_paper_migration(self):
        for task in self.tasks:
            m = random.choice(self.machines)
            m.assign_task(task)

        num_rounds = 40
        for r in range(num_rounds):
            nodes = list(self.machines)
            random.shuffle(nodes)

            for n in nodes:
                m = random.choice(self.machines)
                if n.machine_id == m.machine_id: continue

                level_n = self.alpha * n.sum_mean_time + self.beta * n.sum_var_time
                level_m = self.alpha * m.sum_mean_time + self.beta * m.sum_var_time

                avg_mu = (n.sum_mean_time + m.sum_mean_time) / 2.0
                avg_sigma = (n.sum_var_time + m.sum_var_time) / 2.0
                threshold = self.alpha * avg_mu + self.beta * avg_sigma

                light, heavy = (n, m) if level_n < level_m else (m, n)
                current_light_level = self.alpha * light.sum_mean_time + self.beta * light.sum_var_time

                if current_light_level < threshold:
                    heavy.tasks.sort(key=lambda t: self.alpha * t.mean_time + self.beta * t.var_time)
                    tasks_to_move = []
                    for t in heavy.tasks:
                        if current_light_level >= threshold: break
                        tasks_to_move.append(t)
                        current_light_level += self.alpha * t.mean_time + self.beta * t.var_time

                    for t in tasks_to_move:
                        heavy.remove_task(t)
                        light.assign_task(t)

        assignment_map = []
        for m in self.machines:
            for t in m.tasks: assignment_map.append((t.task_id, m.machine_id))
        assignment_map.sort(key=lambda x: x[0])
        return assignment_map

    def run(self):
        if self.strategy == 'paper_migration': return self._run_paper_migration()

        assignment_map = []
        for task in self.tasks:
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            else:
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))
        return assignment_map

# ==========================================
# 3. GPU 蒙地卡羅驗證與繪圖
# ==========================================
def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}

    if USE_GPU:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    total_tasks = len(task_pool)

    print(f"\n🚀 開始 Monte Carlo 驗證 (Runs={num_runs})...")
    print(f"{'Strategy':<25} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 55)

    for name, config in strategies_config.items():
        start_time = time.time()
        sim = Simulator(num_machines, strategy=config['strategy'], pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = total_tasks / elapsed if elapsed > 0 else 0
        print(f"{name:<25} | {elapsed:<10.4f} | {fps:<10.0f}")

        # --- GPU 驗證邏輯 ---
        if USE_GPU:
            assign_matrix = cp.zeros((total_tasks, num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)

            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, total_tasks)).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)

                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)

                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load, 'assign_map': assign_map}

    return results

def plot_cdf_and_scatter(results, mean_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        axes[0].plot(norm, np.arange(1, len(norm) + 1) / len(norm), label=name, lw=2.5 if 'Paper' in name else 1.5)
    axes[0].set_title('Load Balance CDF'); axes[0].legend(fontsize='small'); axes[0].grid(True, alpha=0.3)

    base_ref_key = 'Baseline (Least Mean)' if 'Baseline (Least Mean)' in results else list(results.keys())[0]
    base_var_val = np.mean(results[base_ref_key]['var_loads'])
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['var_loads']) / (base_var_val if base_var_val > 0 else 1.0)
        axes[1].plot(norm, np.arange(1, len(norm) + 1) / len(norm), label=name, lw=2.5 if 'Paper' in name else 1.5)
    axes[1].set_title('Risk CDF'); axes[1].legend(fontsize='small'); axes[1].grid(True, alpha=0.3)

    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        axes[2].scatter(cv_load, cv_risk, s=250 if 'Paper' in name else 100, label=name, marker='*' if 'Paper' in name else 'o')
    axes[2].set_title('CV Trade-off'); axes[2].legend(fontsize='x-small'); axes[2].grid(True, alpha=0.5)
    plt.tight_layout(); plt.show()

def plot_machine_load_bars(results, mean_times, num_machines):
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()
    target_load = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for i, name in enumerate(sorted_keys):
        ax = axes[i]
        assign_map = results[name]['assign_map']
        machine_tasks = {m: [] for m in range(num_machines)}
        for t_id, m_id in assign_map: machine_tasks[m_id].append(mean_times[t_id])

        # 先計算各機器總負載，用於設定 Y 軸上限
        current_machine_loads = [sum(tasks) for tasks in machine_tasks.values()]
        max_load = max(current_machine_loads) if current_machine_loads else target_load

        for m_id in range(num_machines):
            bottom = 0.0
            for task_load in machine_tasks[m_id]:
                ax.bar(m_id, task_load, bottom=bottom, color='skyblue', edgecolor='black', linewidth=0.5, alpha=0.9)
                bottom += task_load
        ax.axhline(target_load, color='red', linestyle='--', linewidth=2, label=f'Target Average: {target_load:.2f}')
        ax.set_title(f"Method: {name}", fontsize=14, fontweight='bold')
        ax.set_xlabel('Machine ID'); ax.set_ylabel('Total Mean Load')
        ax.set_ylim(0, max_load * 1.15)
        ax.legend(loc='upper right'); ax.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout(); plt.show()

# ==========================================
# 4. 主程式
# ==========================================
if __name__ == "__main__":
    np.random.seed(42); random.seed(42)
    print("1. 讀取 google.csv 數據...")
    try:
        df = pd.read_csv('/content/google.csv')
        mean_times = df['average_usage_mean'].values
        var_times = df['average_usage_variance'].values
        print(f"成功讀取 {len(df)} 筆任務數據")
    except Exception as e:
        print(f"讀取失敗: {e}")
        mean_times = np.random.uniform(10, 100, 5000)
        var_times = np.random.gamma(2.0, 15.0, 5000)

    NUM_MACHINES = 100
    NUM_RUNS = 10000
    strategies_config = {
        'Baseline (Least Mean)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}},
        'Paper Algo (̑\u03b1=1, \u03b2=0)': {'strategy': 'paper_migration', 'params': {'alpha': 1, 'beta': 0}},
        'Paper Algo (\u03b1=0, \u03b2=1)': {'strategy': 'paper_migration', 'params': {'alpha': 0, 'beta': 1}}
    }
    results = run_evaluation_comparison(mean_times, var_times, strategies_config, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, NUM_MACHINES)
    plot_machine_load_bars(results, mean_times, NUM_MACHINES)
```

```python
import random
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# --- GPU 加速檢查 (用於 MC 驗證) ---
try:
    import cupy as cp
    print(f"✅ GPU 加速已啟用: {cp.cuda.runtime.getDeviceCount()} GPUs (用於 MC 驗證)")
    mempool = cp.get_default_memory_pool()
    USE_GPU = True
except ImportError:
    import numpy as cp
    print("⚠️ 警告: 未檢測到 GPU，將使用 CPU 執行")
    USE_GPU = False

# ==========================================
# 1. 資料結構
# ==========================================
class Task:
    __slots__ = ['task_id', 'mean_time', 'var_time', 'actual_time']

    def __init__(self, task_id, mean_time, var_time):
        self.task_id = int(task_id)
        self.mean_time = float(mean_time)
        self.var_time = float(var_time)
        # 靜態派發階段以 mean_time 為基準
        self.actual_time = self.mean_time

class Machine:
    __slots__ = ['machine_id', 'total_load', 'sum_mean_time', 'sum_var_time', 'tasks']

    def __init__(self, machine_id):
        self.machine_id = int(machine_id)
        self.total_load = 0.0
        self.sum_mean_time = 0.0
        self.sum_var_time = 0.0
        self.tasks = []  # 紀錄該機器目前持有的所有任務

    def assign_task(self, task):
        self.total_load += task.actual_time
        self.sum_mean_time += task.mean_time
        self.sum_var_time += task.var_time
        self.tasks.append(task)

    def remove_task(self, task):
        self.total_load -= task.actual_time
        self.sum_mean_time -= task.mean_time
        self.sum_var_time -= task.var_time
        self.tasks.remove(task)

# ==========================================
# 2. 模擬器 (包含論文去中心化演算法)
# ==========================================
class Simulator:
    def __init__(self, num_machines, strategy='least_mu', pre_generated_tasks=None, **kwargs):
        self.num_machines = num_machines
        self.machines = [Machine(i) for i in range(num_machines)]
        self.strategy = strategy
        self.tasks = pre_generated_tasks if pre_generated_tasks else []

        # 論文演算法參數
        self.alpha = kwargs.get('alpha', 1.0)
        self.beta = kwargs.get('beta', 1.0)

        # Yu's 演算法參數
        self.yu_capacity_multiplier = kwargs.get('yu_capacity_multiplier', 2.4)
        self.yu_prob_threshold = kwargs.get('yu_prob_threshold', 0.95)

        # === 新增：Fast Math 演算法相關變數 ===
        self.global_sum_load = 0.0
        self.global_sum_sq_load = 0.0
        self.global_sum_var = 0.0  # <--- 新增：追蹤已分配任務的總變異數
        self.sigma_dispatch_count = 0
        self.fallback_count = 0

    def _run_paper_migration(self):
        """實作論文：初始隨機放置 + 去中心化多回合任務轉移"""
        # 1. 初始隨機放置
        for task in self.tasks:
            m = random.choice(self.machines)
            m.assign_task(task)

        # 2. 回合制任務轉移 (Round-based Execution)
        num_rounds = 10000
        for r in range(num_rounds):
            nodes = list(self.machines)
            random.shuffle(nodes)

            for n in nodes:
                # Algorithm 1: 均勻獨立抽樣
                m = random.choice(self.machines)
                if n.machine_id == m.machine_id:
                    continue

                level_n = self.alpha * n.sum_mean_time + self.beta * n.sum_var_time
                level_m = self.alpha * m.sum_mean_time + self.beta * m.sum_var_time

                avg_mu = (n.sum_mean_time + m.sum_mean_time) / 2.0
                avg_sigma = (n.sum_var_time + m.sum_var_time) / 2.0
                threshold = self.alpha * avg_mu + self.beta * avg_sigma

                light, heavy = (n, m) if level_n < level_m else (m, n)
                current_light_level = self.alpha * light.sum_mean_time + self.beta * light.sum_var_time

                # Algorithm 2: 排序與搬移
                if current_light_level < threshold:
                    heavy.tasks.sort(key=lambda t: self.alpha * t.mean_time + self.beta * t.var_time)

                    tasks_to_move = []
                    for t in heavy.tasks:
                        if current_light_level >= threshold:
                            break
                        tasks_to_move.append(t)
                        current_light_level += self.alpha * t.mean_time + self.beta * t.var_time

                    for t in tasks_to_move:
                        heavy.remove_task(t)
                        light.assign_task(t)

        assignment_map = []
        for m in self.machines:
            for t in m.tasks:
                assignment_map.append((t.task_id, m.machine_id))
        assignment_map.sort(key=lambda x: x[0])
        return assignment_map

    # --- 新增：Yu's 預測型演算法 (適應為 Online Dispatch) ---
    def _dispatch_yu(self, task):
        # 1. 預先計算全系統的極限容量 C (僅在第一次呼叫時計算)
        if not hasattr(self, 'capacity_limit'):
            total_expected_load = sum(t.mean_time for t in self.tasks)
            self.A_mu = total_expected_load / self.num_machines
            # 使用自定義的參數倍數 (預設 2.4)
            self.capacity_limit = self.yu_capacity_multiplier * self.A_mu

        task_mu = task.mean_time
        task_var = task.var_time

        best_m = None
        best_prob = float('inf')

        # 2. 尋找放入任務後，超載機率 (P(X > Capacity)) 最小的機器
        for m in self.machines:
            new_mu = m.sum_mean_time + task_mu
            new_var = m.sum_var_time + task_var
            new_std = new_var ** 0.5 if new_var > 0 else 1e-9

            # 計算標準化分數 Z-score
            z = (self.capacity_limit - new_mu) / new_std

            # 轉換為突破容量上限的實際機率 P(X > C)
            # 使用 math.erf (誤差函數) 來計算常態分佈的右尾機率 (Survival Function)
            prob_exceed = 0.5 * (1.0 - math.erf(z / 1.41421356237))

            if prob_exceed < best_prob:
                best_prob = prob_exceed
                best_m = m

        # (選配) 使用 95% 閾值作為預警：
        # 如果連最好的機器，其超載機率都已經突破了使用者設定的閾值 (例如 0.95)，
        # 代表叢集資源極度緊繃，此時 Yu's 演算法在真實系統中會強制拒絕或發出嚴重警告。
        if best_prob > self.yu_prob_threshold and not hasattr(self, 'warned_hotspot'):
            print(f"\n[Yu's 預警] 系統已飽和！最佳目標機器的超載機率 ({best_prob:.1%}) 已突破設定的熱點閾值 ({self.yu_prob_threshold:.1%})")
            self.warned_hotspot = True

        return best_m

    def _predict_var_fast(self, machine, task_load):
        n = self.num_machines
        # 修改：原本使用 total_load，現在使用 sum_mean_time
        old_load = machine.sum_mean_time
        new_load = old_load + task_load

        predicted_sum_sq = self.global_sum_sq_load - (old_load**2) + (new_load**2)
        predicted_sum = self.global_sum_load + task_load

        predicted_mean = predicted_sum / n
        predicted_var = (predicted_sum_sq / n) - (predicted_mean**2)
        return predicted_var

    def _dispatch_fast_math(self, task):
        task_load = task.mean_time
        task_var = task.var_time  # 取得當前任務的變異數

        # 1. 取得所有機器的當前負載 (sum_mean_time)
        loads = [m.sum_mean_time for m in self.machines]
        n = len(loads)

        if n > 1:
            avg_load = np.mean(loads)
            var_load = np.var(loads)
            std_load = var_load**0.5
            threshold = avg_load - (2 * std_load)

            # 5. 找出在分佈中「小於 2 倍標準差」的節點
            sigma_candidates = [m for m in self.machines if m.sum_mean_time < threshold]

            # 6. 若有極度閒置的候選人
            if sigma_candidates:
                self.sigma_dispatch_count += 1
                return min(self.machines, key=lambda m: m.sum_mean_time)

        # === 更新：精確的動態變異數閾值過濾 ===
        candidates = []

        # 計算包含當前任務的「預期總變異數」
        current_total_var = self.global_sum_var + task_var
        # 平均分攤到所有機器上的「完美理想變異數」作為嚴格閾值
        dynamic_var_threshold = current_total_var / self.num_machines

        for machine in self.machines:
            # 預測放進去後的變異數，是否小於等於這個完美的理想基準
            if self._predict_var_fast(machine, task_load) <= dynamic_var_threshold:
                candidates.append(machine)

        if not candidates:
            # === 觸發 Fallback (退化為 Least Mu) ===
            self.fallback_count += 1
            return min(self.machines, key=lambda m: m.sum_mean_time)

        # 在及格的候選人中，挑選累積風險最小的
        return min(candidates, key=lambda m: m.sum_var_time)

    def run(self):
        if self.strategy == 'paper_migration':
            return self._run_paper_migration()

        assignment_map = []
        for task in self.tasks:
            # 決定派發目標機器
            if self.strategy == 'least_mu':
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)
            elif self.strategy == 'least_var':
                best_m = min(self.machines, key=lambda m: m.sum_var_time)
            elif self.strategy == 'yu':
                best_m = self._dispatch_yu(task)
            elif self.strategy == 'fast_math':
                best_m = self._dispatch_fast_math(task)
            else:
                best_m = min(self.machines, key=lambda m: m.sum_mean_time)

            # --- 更新 Fast Math 需要的全域狀態 ---
            old_load = best_m.sum_mean_time
            new_load = old_load + task.mean_time

            self.global_sum_load += task.mean_time
            self.global_sum_sq_load = self.global_sum_sq_load - (old_load**2) + (new_load**2)
            self.global_sum_var += task.var_time  # <--- 新增：累加已分配任務的變異數
            # ------------------------------------------------

            best_m.assign_task(task)
            assignment_map.append((task.task_id, best_m.machine_id))

        if self.strategy == 'fast_math':
            print(f"[Fast Math] 派發完成: Sigma 觸發={self.sigma_dispatch_count}次, Fallback 觸發={self.fallback_count}次")

        return assignment_map


# ==========================================
# 3. GPU 蒙地卡羅驗證與繪圖
# ==========================================
def run_evaluation_comparison(mean_times, var_times, strategies_config, num_machines, num_runs):
    results = {}

    if USE_GPU:
        means_gpu = cp.array(mean_times, dtype=cp.float32)
        stds_gpu = cp.sqrt(cp.array(var_times, dtype=cp.float32))

    task_pool = [Task(i, m, v) for i, (m, v) in enumerate(zip(mean_times, var_times))]
    total_tasks = len(task_pool)

    print(f"\n🚀 開始 Monte Carlo 驗證 (Runs={num_runs})...")
    print(f"{'Strategy':<25} | {'Time (s)':<10} | {'FPS':<10}")
    print("-" * 55)

    for name, config in strategies_config.items():
        start_time = time.time()
        sim = Simulator(num_machines, strategy=config['strategy'], pre_generated_tasks=task_pool, **config['params'])
        assign_map = sim.run()

        elapsed = time.time() - start_time
        fps = total_tasks / elapsed if elapsed > 0 else 0
        print(f"{name:<25} | {elapsed:<10.4f} | {fps:<10.0f}")

        # --- GPU 驗證邏輯 ---
        if USE_GPU:
            assign_matrix = cp.zeros((total_tasks, num_machines), dtype=cp.float32)
            for t_id, m_id in assign_map:
                assign_matrix[t_id, m_id] = 1.0

            sum_load = cp.zeros(num_machines, dtype=cp.float32)
            sum_sq_load = cp.zeros(num_machines, dtype=cp.float32)

            BATCH = 2000
            for _ in range(0, num_runs, BATCH):
                current_batch = min(BATCH, num_runs - _)
                batch_rand = cp.random.normal(means_gpu, stds_gpu, size=(current_batch, total_tasks)).astype(cp.float32)
                batch_rand = cp.maximum(0.1, batch_rand)
                batch_loads = cp.dot(batch_rand, assign_matrix)

                sum_load += cp.sum(batch_loads, axis=0)
                sum_sq_load += cp.sum(batch_loads ** 2, axis=0)

                del batch_rand, batch_loads
                try: mempool.free_all_blocks()
                except: pass

            avg_load_gpu = sum_load / num_runs
            var_load_gpu = (sum_sq_load / num_runs) - (avg_load_gpu ** 2)
            avg_load = cp.asnumpy(avg_load_gpu)
            var_load = cp.asnumpy(var_load_gpu)
        else:
            m_means = np.zeros(num_machines)
            m_vars = np.zeros(num_machines)
            for t_id, m_id in assign_map:
                m_means[m_id] += mean_times[t_id]
                m_vars[m_id] += var_times[t_id]
            avg_load = m_means
            var_load = m_vars

        results[name] = {'avg_loads': avg_load, 'var_loads': var_load}

    return results

def plot_cdf_and_scatter(results, mean_times, var_times, num_machines):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Load CDF
    ax = axes[0]
    base_mu = sum(mean_times) / num_machines
    sorted_keys = sorted(results.keys())
    for name in sorted_keys:
        data = results[name]
        norm = np.sort(data['avg_loads']) / base_mu
        y = np.arange(1, len(norm)+1) / len(norm)

        # --- 修改：為不同演算法設定專屬的線條粗細與圖層順序 ---
        if 'Fast Math' in name:
            lw = 3.5      # 特別加粗您的演算法
            zorder = 10   # 畫在最上層，避免被覆蓋
            color = 'red'
        elif 'Paper' in name:
            lw = 1.5
            zorder = 1
            color = None
        else:
            lw = 1.5
            zorder = 1
            color = None

        ax.plot(norm, y, label=name, lw=lw, zorder=zorder, color=color)

    ax.set_title('Load Balance CDF (Steeper is better)')
    ax.set_xlabel('Normalized Load')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 2. Risk CDF
    ax = axes[1]
    base_var = sum(var_times) / num_machines

    for name in sorted_keys:
        data = results[name]
        denominator = base_var if base_var > 0 else 1.0
        norm = np.sort(data['var_loads']) / denominator
        y = np.arange(1, len(norm)+1) / len(norm)

        # --- 修改：為不同演算法設定專屬的線條粗細與圖層順序 ---
        if 'Fast Math' in name:
            lw = 3.5      # 特別加粗您的演算法
            zorder = 10    # 畫在最上層
            color = 'red'
        elif 'Paper' in name:
            lw = 1.5
            zorder = 1
            color = None
        else:
            lw = 1.5
            zorder = 1
            color = None

        ax.plot(norm, y, label=name, lw=lw, zorder=zorder, color=color)

    ax.set_title('Risk CDF (Normalized Variance)')
    ax.set_xlabel('Normalized Variance (Multiple of Ideal Var)')
    ax.legend(fontsize='small')
    ax.grid(True, alpha=0.3)

    # 3. Pareto Scatter
    ax = axes[2]
    for name in sorted_keys:
        data = results[name]
        cv_load = np.std(data['avg_loads']) / np.mean(data['avg_loads'])
        cv_risk = np.std(data['var_loads']) / np.mean(data['var_loads'])
        total_cost = cv_load + cv_risk

        # 預設標記
        marker = 'o'
        size = 100
        zorder = 1
        color = None

        # --- 修改：設定散佈圖的專屬標記 ---
        if 'Baseline' in name:
            marker = 'X'
            size = 150
        elif 'Paper' in name:
            marker = '*'
            size = 150
            zorder = 5
        elif 'Fast Math' in name:
            marker = 'D'  # D 代表鑽石 (Diamond) 形狀
            size = 300    # 最大最醒目
            zorder = 5   # 確保顯示在最上層
            color = 'red'

        ax.scatter(cv_load, cv_risk, s=size, label=f"{name}\n(Cost={total_cost:.3f})",
                   edgecolors='k', marker=marker, zorder=zorder, color=color)

        offset = 0.001 if 'Baseline' in name else -0.001
        ax.text(cv_load, cv_risk + offset, name, fontsize=9, ha='center', zorder=zorder)

    ax.set_title('Performance Trade-off (Left-Bottom is Better)')
    ax.set_xlabel('Load Imbalance (CV)')
    ax.set_ylabel('Risk Imbalance (CV)')
    ax.legend(fontsize='x-small', loc='best')
    ax.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# ==========================================
# 4. 主程式配置與執行
# ==========================================
if __name__ == "__main__":
    np.random.seed(42)
    random.seed(42)

    NUM_MACHINES = 100
    NUM_TASKS = 10000
    NUM_RUNS = 10000

    print("1. 讀取 google.csv 數據...")
    try:
      raise Exception("強制進入極端數據測試模式！")
      # df = pd.read_csv('/content/google.csv')
      # mean_times = df['average_usage_mean'].values
      # var_times = df['average_usage_variance'].values
      # df = pd.read_csv('/content/azure.csv')
      # mean_times = df['vm_load_mean_mips'].values
      # var_times = df['vm_load_std_mips'].values
      # df = pd.read_csv('/content/alibaba.csv')
      # mean_times = df['cpu_mean'].values
      # var_times = df['cpu_variance'].values
      print(f"成功讀取 {len(df)} 筆任務數據")
    except Exception as e:
      print(f"讀取失敗: {e}")
      NUM_TASKS = 10000
      EXTREME_RATIO = 0.05  # 設定 5% 為極端地雷任務

      num_normal = int(NUM_TASKS * (1 - EXTREME_RATIO))
      num_extreme = NUM_TASKS - num_normal

      # 1. 生成 95% 的常規平穩任務 (老鼠)
      normal_means = np.random.uniform(10, 100, num_normal)
      # 常規 Gamma 分佈的變異數大約在 30 左右 (2.0 * 15.0)
      normal_vars = np.random.gamma(2.0, 15.0, num_normal)

      # 2. 生成 5% 的極端波動任務 (大象)
      # 稍微拉高一點平均負載，讓 GLM 覺得它們看起來「還算普通」
      extreme_means = np.random.uniform(50, 150, num_extreme)
      # 致命武器：使用 Pareto 分佈 (重尾分佈) 產生極端變異數，並放大 100 倍！
      # 這些任務的變異數可能會飆到數千甚至上萬，模擬系統崩潰邊緣的突發雜訊
      extreme_vars = np.random.pareto(1.5, num_extreme) * 1000

      # 3. 合併並打亂陣列 (避免極端任務全集中在最後)
      mean_times = np.concatenate([normal_means, extreme_means])
      var_times = np.concatenate([normal_vars, extreme_vars])

      indices = np.arange(NUM_TASKS)
      np.random.shuffle(indices)

      mean_times = mean_times[indices]
      var_times = var_times[indices]

      print(f"數據生成完畢：包含 {num_normal} 筆常規任務與 {num_extreme} 筆極端任務。")

    TOP_K = 10000

    # np.argsort 會回傳從小到大的索引，使用 [::-1] 將其反轉為由大到小，並取前 TOP_K 筆
    top_var_indices = np.argsort(var_times)[::-1][:TOP_K]

    # 利用這些索引重新過濾陣列
    mean_times = mean_times[top_var_indices]
    var_times = var_times[top_var_indices]

    NUM_TASKS = len(mean_times)
    print(f"🔥 壓力測試準備就緒：已篩選出變異數最大的前 {NUM_TASKS} 筆任務！")

    strategies_config = {
        'Baseline (Least Mean)': {'strategy': 'least_mu', 'params': {}},
        'Baseline (Least Var)': {'strategy': 'least_var', 'params': {}},
        "Hsiao's (α=1, β=0)": {'strategy': 'paper_migration', 'params': {'alpha': 1.0, 'beta': 0}},
        "Hsiao's (α=0, β=1)": {'strategy': 'paper_migration', 'params': {'alpha': 0, 'beta': 1.0}},
        # "Hsiao's (α=0.9, β=0.1)": {'strategy': 'paper_migration', 'params': {'alpha': 0.9, 'beta': 0.1}},
        # "Yu's (Predictive)": {'strategy': 'yu', 'params': {'yu_capacity_multiplier': 2.4, 'yu_prob_threshold': 0.95}},

        # === 新增：您的自訂演算法 ===
        'My Fast Math Algo': {
            'strategy': 'fast_math',
            'params': {}  # <--- 直接留空即可，因為演算法已經會自己計算動態完美極限了！
        }
    }
    results = run_evaluation_comparison(mean_times, var_times, strategies_config, NUM_MACHINES, NUM_RUNS)
    plot_cdf_and_scatter(results, mean_times, var_times, NUM_MACHINES)
```
